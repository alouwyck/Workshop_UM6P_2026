# **Part 7 - Linear Programming for Groundwater Management**

## Introduction

Linear programming is a **mathematical optimization technique** used to find the best outcome—such as minimizing costs or maximizing efficiency—subject to a set of constraints. In a linear programming problem, both the **objective function** (what you want to optimize) and the **constraints** (the limitations or requirements of the system) are expressed as **linear equations or inequalities**. This makes it possible to solve complex decision-making problems in a structured and computationally efficient way.

In the context of **groundwater management**, linear programming is particularly useful for designing and controlling pumping and reinjection systems. As discussed in the previous notebook, **groundwater systems** are governed by physical laws (such as conservation of mass and Darcy's law), which can often be approximated or reformulated into **linear relationships**. By defining variables such as pumping rates at wells and reinjection rates at recharge locations, one can construct a model where the objective is, for example, to minimize the total amount of water pumped while still satisfying operational and environmental constraints.

For systems involving **combined pumping and reinjection**, an important condition is that the total extraction and reinjection rates balance out, meaning the **net withdrawal** from the aquifer **is zero**. This constraint ensures long-term sustainability and prevents depletion of the groundwater resource. Linear programming allows us to determine the **optimal distribution of** pumping and reinjection **rates** across multiple wells, while respecting constraints such as maintaining hydraulic heads within safe limits, avoiding contamination spread, or meeting water demand.

## Mathematical Formulation

The mathematical formulation of a linear programming problem consists of three main components:
- **decision variables**,
- an **objective function**,
- and a set of **constraints**.

First, we define the **decision variables**. In the context of groundwater management, these variables typically represent the **pumping and reinjection rates** at different wells.

Let $Q_i$ denote the rate [L³/T] at well $i$, where positive values correspond to pumping (extraction) and negative values correspond to reinjection. If there are $n$ wells in the system, the decision variables can be collected into a vector:

> $ \mathbf{q} = (Q_1, Q_2, \dots, Q_n) $.

Next, we define the **objective function**. In this case, the goal is to **minimize the total pumping rate**. Since reinjection corresponds to negative values of $Q_i$, the total pumping can be expressed as the sum of the positive components of $\mathbf{q}$. In a linear programming framework, this is often approximated or reformulated as a linear function, for example:

> $\min \sum\limits_{i=1}^{n} c_i Q_i$

where $c_i$ are coefficients that weight the contribution of each well $i$. In the simplest case, we define these coefficients as:

> $
c_i =
\begin{cases}
1 & \text{if well } i \text{ is a pumping well} \\
0 & \text{if well } i \text{ is a reinjection well}
\end{cases}
$


The system is subject to several **constraints**. A key physical constraint is the **balance between pumping and reinjection**:
> $\sum\limits_{i=1}^{n} Q_i = 0$

which ensures that the total extraction is exactly compensated by total reinjection, maintaining a steady-state condition in the aquifer.

**Additional constraints** arise from hydraulic considerations. The hydraulic head $h$ [L] or **drawdown** $s$ [L] at various control points in the aquifer must remain within specified bounds to prevent issues such as excessive drawdown or unwanted spreading of contaminants. These constraints can often be expressed as linear relationships between the heads and the pumping rates:

> $ s_j = \sum\limits_{i=1}^{n} u_{ji} Q_i$

where $s_j$ is the drawdown at location $j$, and $u_{ji}$ are influence coefficients [L/L³] that describe how pumping at well $i$ affects the drawdown at location $j$. If the principle of superposition is applied, then $u_{ji}$ represents the drawdown at location $j$ caused by unit pumping of well $i$ (i.e., $Q_i = 1$). Using the Thiem formula, this gives:

> $u_{ji} = \dfrac{\ln r_{ji}}{2\pi T}$

with $r_{ji}$ the Euclidean distance [L] between location $j$ and well $i$. Recall that the radius of influence $R$ [L] cancels out due to the constraint that balances pumping and reinjection.

Imposing upper and lower bounds on $s_j$ leads to linear inequality constraints:
> $s_j^{\text{min}} \leq \sum\limits_{i=1}^{n} u_{ji} Q_i \leq s_j^{\text{max}}$

Finally, **operational constraints** may also be included, such as limits on the capacity of individual wells:

> $Q_i^{\text{min}} \leq Q_i \leq Q_i^{\text{max}} \quad \text{for } i = 1, \dots, n$

Putting everything together, the general linear programming formulation can be written compactly as:

> $\min_{\mathbf{q}} \ \mathbf{c}^\top \mathbf{q}$

subject to:

> $\mathbf{U} \mathbf{q} \leq \mathbf{s}, \quad \mathbf{U}_{\text{eq}} \mathbf{q} = \mathbf{s}_{\text{eq}}, \quad \mathbf{q}^{\text{min}} \leq \mathbf{q} \leq \mathbf{q}^{\text{max}}$

If there are $m$ locations $j$, then $\mathbf{U}$ is an $m$ x $n$ matrix containing the coefficients $u_{ji}$. Vectors $\mathbf{s}$ and $\mathbf{s}_{\text{eq}}$ hold the drawdown limits $s_j^{\text{min}}$ and/or $s_j^{\text{max}}$.

This mathematical formulation provides a powerful way to model groundwater management problems, allowing the incorporation of groundwater well-flow, environmental constraints, and operational limits into a single optimization framework. Let’s now work through a few code examples to show how linear programming can be easily applied in practice.


## OO Python

In this section, we first prepare the code that we'll need. We import the `numpy` library and the `pyplot` module, and define the `thiem` function as well as the `Well` and `Model` classes that we implemented in the previous notebook.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def thiem(Q, T, R, r):
    '''
    Simulate drawdown using the Thiem equation
    Q: pumping rate [L³/T] - float
    T: aquifer transmissivity [L²/T] - float
    R: radius of influence [L] - float
    r: radial distance [L] - array-like
    returns drawdown s [L] for each distance r
    '''
    return Q/2/np.pi/T * np.log(np.array(r)/R)
```

```python
class Well:
    """
    A pumping or injection well represented by the Thiem equation.

    Parameters
    ----------
    model : Model
        The Model object to which the well belongs.
    Q : float
        Pumping rate [L³/T]. Positive for pumping, negative for injection.
    xw, yw : float
        Coordinates [L] of the well.
    rw : float, optional
        Well radius [L]. Distances smaller than this are set to rw
        to avoid singularities. Default is 0.1 m.
    """

    def __init__(self, model, Q, xw, yw, rw=0.1):
        self.model = model
        self.Q = Q
        self.xw, self.yw = xw, yw
        self.rw = rw

    def s(self, x, y):
        """
        Compute drawdown at coordinates (x, y) due to this well
        using the Thiem equation.

        Parameters
        ----------
        x, y : float or array-like
            Coordinates [L] at which the drawdown is evaluated.

        Returns
        -------
        ndarray or float
            Drawdown value(s) [L] at the given coordinate(s).
        """
        x, y = np.array(x), np.array(y)
        r = np.sqrt((x - self.xw)**2 + (y - self.yw)**2)
        if x.ndim: r[r < self.rw] = self.rw  # prevent singularity
        elif r < self.rw: r = self.rw

        return thiem(self.Q, self.model.T, self.model.R, r)
```

```python
class Model:
    """
    Model for single homogeneous aquifer represented by transmissivity, initial head,
    and a list of wells.

    Parameters
    ----------
    T : float
        Aquifer transmissivity [L²/T].
    h0 : float, optional
        Initial hydraulic head [L] in the aquifer. Default is 0.
    R : float, optional
        Approximate distance [L] to constant-head point for wells in the aquifer.
        Default is 1e6.
    """

    def __init__(self, T, h0=0, R=1e6):
        self.T = T
        self.h0 = h0
        self.R = R
        self.wells = []

    def add_well(self, Q, xw, yw, rw=0.1):
        """
        Add a pumping or injection well to the model.

        Parameters
        ----------
        Q : float
            Pumping rate [L³/T].
        xw, yw : float
            Well coordinates [L].
        rw : float, optional, default is 0.1
            Well radius [L].

        Returns
        -------
        Well
            The newly created Well object.
        """
        well = Well(self, Q, xw, yw, rw)
        self.wells.append(well)
        return well

    def h(self, x, y):
        """
        Compute the hydraulic head at coordinates (x, y),
        obtained by superposition of all wells.

        Parameters
        ----------
        x, y : float or array-like
            Coordinates [L] at which the head is evaluated.

        Returns
        -------
        ndarray or float
            Hydraulic head [L] value(s) at the given coordinate(s).
        """
        return self.h0 + np.sum([well.s(x, y) for well in self.wells], axis=0)
```

## Pumping and Reinjection Well System

Let's start with a simple two-well system:
- the distance between pumping and injection well is 50 m;
- the pumping rate is 3000 m³/d;
- the aquifer has a transmissivity of 3000 m²/d.

```python
T = 100   # aquifer transmissivity (m²/d)
Q = 3000  # pumping rate (m³/d)

# well coordatines
x_pump, y_pump = 25, 0  # pumping well
x_inj, y_inj = -25, 0   # injection well

# grid
x, y = np.meshgrid(np.linspace(-100, 100, 100),  # x-axis: 100 point between -100 and 100
                   np.linspace(-50, 50, 100))    # y-axis: 100 point between -50 and 50
```

Building the model and simulate the drawdowns:

```python
model = Model(T)                          # model
pump = model.add_well(Q, x_pump, y_pump)  # pumping well
inj = model.add_well(-Q, x_inj, y_inj)    # injection well
s = model.h(x, y)                         # simulate drawdowns
```

Visualizing the result using a contour plot:

```python
plt.figure(figsize=(10, 5))
plt.contourf(x, y, s, levels=np.arange(-25, 26, 1))
plt.colorbar()
ax = plt.gca() # Get the current axes
ax.set_aspect('equal') # Set aspect ratio to equal
```

Computing the drawdown at point (10, 0):

```python
print(model.h(10, 0))
```

Suppose we now want the drawdown at point (10, 0) to be equal to -2 m. This is a drawdown constraint and the **analytical solution** to this problem is very straightforward. Applying the principle of superposition to the Thiem equation gives:  

> $s_{\text{target}} = Q_{\text{opt}} \dfrac{\ln r_{\text{pump}} - \ln r_{\text{inj}}}{2\pi T}$

Rearranging this equation yields the pumping rate $Q_{\text{opt}}$ at which drawdown $s_{\text{target}}$ at target point (10, 0) is -2 m:

> $Q_{\text{opt}} = \dfrac{2\pi T s_{\text{target}}}{\ln r_{\text{pump}} - \ln r_{\text{inj}}}$

The denominator can also be written as:

> $\ln r_{\text{pump}} - \ln r_{\text{inj}} = \ln \dfrac{r_{\text{pump}}}{r_{\text{inj}}}$

with $r_{\text{pump}}$ and $r_{\text{inj}}$ the Euclidean distances between target point $(x_{\text{target}}, y_{\text{target}})$ and injection and pumping well, respectively:

> $r_{\text{pump}} = \sqrt{\left(x_{\text{target}}-x_{\text{pump}}\right)^2 + \left(y_{\text{target}}-y_{\text{pump}}\right)^2}$

> $r_{\text{inj}} = \sqrt{\left(x_{\text{target}}-x_{\text{inj}}\right)^2 + \left(y_{\text{target}}-y_{\text{inj}}\right)^2}$

Let's code this formula:

```python
s_target = -2
x_target, y_target = 10, 0

r_pump = np.sqrt((x_target - x_pump)**2 + (y_target - y_pump)**2)
r_inj = np.sqrt((x_target - x_inj)**2 + (y_target - y_inj)**2)

Q_opt = 2*np.pi*T*s_target / np.log(r_pump/r_inj)
print('optimal pumping rate:', Q_opt, 'm³/d')
```

The optimal pumping rate for the given constraint is 1483 m³/d. Of course, we can also use simple **proportional scaling**:

```python
s_current = model.h(x_target, y_target)
Q_opt = Q / s_current * s_target
print('optimal pumping rate:', Q_opt, 'm³/d')
```

Let's rebuild the model using the derived optimal pumping rate and verify the constraint:

```python
model_opt = Model(T)                              # model
pump = model_opt.add_well(Q_opt, x_pump, y_pump)  # pumping well
inj = model_opt.add_well(-Q_opt, x_inj, y_inj)    # injection well
s_opt = model_opt.h(x, y)                         # simulate drawdowns
print(model_opt.h(x_target, y_target).item())     # drawdown at target point
```

We get indeed a drawdown of -2 m at point (10, 0). Let's plot the corresponding drawdown map:

```python
plt.figure(figsize=(10, 5))
plt.contourf(x, y, s_opt, levels=np.arange(-12, 13, 1))
plt.colorbar()
plt.plot(x_target, y_target, 'ro')
ax = plt.gca() # Get the current axes
ax.set_aspect('equal') # Set aspect ratio to equal
```

Finally, we apply **linear programming** to solve the problem. In this case, the optimization problem simplifies to:

> objective function:
>    $$\min Q$$
>
> subject to:
>
> $$Q \ge 0$$
>
> $$u \cdot Q = 2$$

with $Q$ the pumping rate [L³/T] and $u$ the drawdown per unit pumping rate [L/L³] at point (10, 0).

We apply `scipy` function `linprog`, which uses the following matrix notation:

> objective function:
> $$\min_\mathbf{x} \mathbf{c}^T \mathbf{x}$$
> subject to:
> $$\mathbf{x} \ge \mathbf{0}$$
> $$\mathbf{A} \mathbf{x} = \mathbf{b}$$

This means that vector $\mathbf{x}$ holds the pumping rate $Q$, vector $\mathbf{c}$ contains a single coefficient equal to 1, matrix $\mathbf{A}$ contains drawdown $u$ per unit pumping rate, and vector $\mathbf{b}$ holds the drawdown constraint equal to -2.

So, we first need to compute $u$, which is done by simulating a model with unit pumping and injection rate:

```python
u_model = Model(T)                   # model
u_model.add_well(1, x_pump, y_pump)  # well with unit pumping rate
u_model.add_well(-1, x_inj, y_inj)   # well with unit injection rate
u = u_model.h(x_target, y_target)    # drawdown caused by unit pumping and reinjection
```

Now we can easily apply function `linprog` to find the optimal pumping rate:

```python
from scipy.optimize import linprog

c = [1]         # 1D vector
A = [[u]]       # 2D matrix
b = [s_target]  # 1D vector
res = linprog(c, A_eq=A, b_eq=b,
              bounds=(0, None))  # constraint: Q must be positive
Q_opt = res.x[0]

print('optimal pumping rate:', Q_opt, 'm³/d')
print('drawdown at point (10, 0):', u * Q_opt, 'm')
```

Note that `linprog()` expects vectors and matrices with specific dimensions:

- `c` is a 1‑D vector → single square brackets: `[1]`
- `b` is a 1‑D vector → `[s_target]`
- `A` is a 2‑D matrix → double brackets: `[[u]]`


## A More Challenging Example

Let's revisit the construction dewatering system that we simulated and optimized in the previous notebook.

### Setup of the construction dewatering system

The example consists of a rectangular excavation of 50 m × 30 m, centered in the model domain. To enable dry working conditions, groundwater levels inside the excavation are lowered using a series of pumping wells placed around the site.

The pumping wells are arranged in a rectangular pattern at a distance of 10 m from the excavation boundary, whereas the distance between two adjacent wells is also 10 m. This creates a controlled drawdown that keeps the groundwater level below the excavation base.

To reduce the impact of dewatering on the surrounding area, the extracted groundwater is re-injected into the aquifer using a second ring of wells located further away. These injection wells are positioned at a distance of 25 m from the excavation boundary, forming a wider perimeter around the site.

The purpose of the reinjection is to compensate for the extracted water, thereby limiting drawdown outside the construction area and maintaining the regional groundwater balance.

```python
# Excavation boundary (brown)
excavation = {"xmin": -25, "xmax": 25, "ymin": -15, "ymax": 15}

# Pumping wells (blue)
pumping_wells = []
for x in np.arange(-25, 30, 10):
    pumping_wells.append((x, 25))
    pumping_wells.append((x, -25))
for y in np.arange(-15, 20, 10):
    pumping_wells.append((-35, y))
    pumping_wells.append((35, y))

# Injection wells (green)
injection_wells = []
for x in [-25, 0, 25]:
    injection_wells.append((x, 40))
    injection_wells.append((x, -40))
for y in [-15, 0, 15]:
    injection_wells.append((-50, y))
    injection_wells.append((50, y))
```

```python
# Function to plot the excavation and the wells
def plot_wells():

    fig, ax = plt.subplots(figsize=(6,6))

    # Excavation
    rect = plt.Rectangle(
        (excavation["xmin"], excavation["ymin"]),
        excavation["xmax"] - excavation["xmin"],
        excavation["ymax"] - excavation["ymin"],
        color="brown"
    )
    ax.add_patch(rect)

    # Wells
    px, py = zip(*pumping_wells)
    ix, iy = zip(*injection_wells)

    ax.scatter(px, py, color="blue")   # pumping
    ax.scatter(ix, iy, color="green")   # injection

    # Limits
    xmin, xmax = -60, 60
    ymin, ymax = -50, 50
    ax.set_xlim(xmin, xmax)
    ax.set_ylim(ymin, ymax)

    # ---- TICKS ----
    # Major ticks every 10 m (with labels)
    ax.set_xticks(np.arange(xmin, xmax + 10, 10))
    ax.set_yticks(np.arange(ymin, ymax + 10, 10))

    # Minor ticks every 5 m (no labels)
    ax.set_xticks(np.arange(xmin, xmax + 5, 5), minor=True)
    ax.set_yticks(np.arange(ymin, ymax + 5, 5), minor=True)

    # ---- GRID ----
    ax.grid(which='minor', linewidth=0.5)
    ax.grid(which='major', linewidth=1.0)

    # Keep square cells
    ax.set_aspect('equal')

    # Return axes
    return ax

# Plot
plot_wells();
```

We consider a homogeneous aquifer with constant thickness equal to 20 m and horizontal conductivity equal to 10 m/d:

```python
T = 20 * 10  # transmissivity T = KD
```

Recall that we need to simulate a model with unit pumping rates in order to apply linear programming. The corresponding injection rates are obtained by balancing them with these pumping rates. Since each pumping well extracts at the same rate and each injection well reinjects at the same rate, the balance equation simplifies to:

> $n_{\text{pump}} \cdot Q_{\text{pump}} = n_{\text{inj}} \cdot -Q_{\text{inj}}$

```python
# number of pumping and injection wells
n_pump, n_inj = len(pumping_wells), len(injection_wells)

Q_pump = 1  # unit pumping rate
Q_inj = -n_pump * Q_pump / n_inj  # corresponding injection rates

print('pumping rate:', Q_pump)
print('injection rate:', Q_inj)
```

Now we're ready to define the superposition model and add the wells:

```python
u_model = Model(T)
for pumping_well in pumping_wells:
    u_model.add_well(Q_pump, *pumping_well)
for injection_well in injection_wells:
    u_model.add_well(Q_inj, *injection_well)
```

### Optimizing the pumping rate

As demonstrated in the previous notebook, the linearity of the model allows us to use simple **proportional scaling** if we want to impose a head-change target (or constraint) at only a single point.

For example, suppose we want to obtain a drawdown of -2 m at the center of the excavation:

```python
x_target, y_target = 0, 0                 # center is at (0, 0)
s_target = -2                             # target drawdown at (0, 0)
u = u_model.h(x_target, y_target).item()  # drawdown at (0, 0) caused by unit pumping

Q_pump = s_target / u             # optimal pumping rate
Q_inj = -n_pump * Q_pump / n_inj  # corresponding injection rate

print('optimal pumping rate:', Q_pump, 'm³/d')
print('optimal injection rate:', Q_inj, 'm³/d')
```

Applying **linear programming** gives the same result:

```python
c = [1]
A = [[u]]
b = [s_target]
res = linprog(c, A_eq=A, b_eq=b, bounds=(0, None))

Q_pump = res.x[0]  # optimal pumping rate
Q_inj = -n_pump * Q_pump / n_inj  # corresponding injection rate

print('optimal pumping rate:', Q_pump, 'm³/d')
print('optimal injection rate:', Q_inj, 'm³/d')
print('drawdown at center:', u * Q_pump, 'm')
```

Let's build a model with these optimal pumping and injection rates:

```python
model_opt = Model(T)
for pumping_well in pumping_wells:
    model_opt.add_well(Q_pump, *pumping_well)
for injection_well in injection_wells:
    model_opt.add_well(Q_inj, *injection_well)
```

Plotting the north–south and east–west cross‑sections of the head change along the x‑ and y‑axes:

```python
cs = np.linspace(-100, 100, 1000)
z = np.zeros(len(cs))
plt.figure(figsize=(12, 4))
plt.plot(cs, model_opt.h(cs, z), "b-", label="east-west")
plt.plot(cs, model_opt.h(z, cs), "k-", label="north-south")
plt.legend()
plt.grid()
plt.xlabel("distance (m)")
plt.ylabel("head change (m)");
```

The drawdown at the center is indeed -2 m. Let’s perform one final verification:

```python
print('drawdown at excavation center:', model_opt.h(x_target, y_target).item(), 'm')
```

### Additional constraint

Suppose we define an **additional head change constraint** for the injections wells: the increase in head at these wells must be at least 1.5 m.

We define the problem mathematically:

> objective function:
> $$\min Q$$
> subject to:
> $$Q \ge 0$$
> at the center of the excavation center:
> $$u_1 \cdot Q \le -2$$
> at the injection wells:
> $$u_2 \cdot Q > 1.5$$

The `linprog` function requires the constraints to be written as ‘less‑than’ inequalities, which is the so‑called canonical form. Therefore, we need to reformulate the third constraint at the injection wells:

> $$-u_2 \cdot Q \le -1.5$$


Using the notation from the `linprog` function documentation, the problem is written as follows:

> objective function:
> $$\min_\mathbf{x} \mathbf{c}^T \mathbf{x}$$
> subject to:
> $$\mathbf{x} \ge \mathbf{0}$$
>
> $$\mathbf{A}_{\text{ub}} \mathbf{x} \le \mathbf{b}_{\text{ub}}$$

First, we define target coordinates and head change constraints, and simulate the head changes caused by unit pumping:

```python
# coordatines of target points
x_center, y_center = 0, 0
x_inj, y_inj       = 0, 40

# head change constraints
s_center = -2
s_inj = 1.5

# head changes corresponding to unit pumping
u_center = u_model.h(x_center, y_center)
u_inj = u_model.h(x_inj, y_inj)
```

Note that we consider only the injection well where the head change is minimal according to the cross-sectional plot.

Now, we apply linear programming:

```python
c = [1]
A = [[u_center], [-u_inj]]
b = [s_center, -s_inj]
res = linprog(c, A_ub=A, b_ub=b, bounds=(0, None))

Q_pump = res.x[0]  # optimal pumping rate
Q_inj = -n_pump * Q_pump / n_inj  # corresponding injection rate

print('optimal pumping rate:', Q_pump, 'm³/d')
print('optimal injection rate:', Q_inj, 'm³/d')
```

Building a model with these optimal rates:

```python
model_opt = Model(T)
for pumping_well in pumping_wells:
    model_opt.add_well(Q_pump, *pumping_well)
for injection_well in injection_wells:
    model_opt.add_well(Q_inj, *injection_well)
```

Verifying the head change constraints:

```python
print('head change at excavation center:', model_opt.h(x_center, y_center), 'm')
print('head change at injection well:', model_opt.h(x_inj, y_inj), 'm')
```

We see that the drawdown at the center of the excavation must be less than -2 m in order to satisfy the head change constraint of 1.5 m at the injection well.

Replotting the cross-sections:

```python
cs = np.linspace(-100, 100, 1000)
z = np.zeros(len(cs))
plt.figure(figsize=(12, 4))
plt.plot(cs, model_opt.h(cs, z), "b-", label="east-west")
plt.plot(cs, model_opt.h(z, cs), "k-", label="north-south")
plt.legend()
plt.grid()
plt.xlabel("distance (m)")
plt.ylabel("head change (m)");
```

### Infeasible problem

When we take a closer look at the cross‑sectional plot, we see that in the North–South section there is still a negative drawdown at 100 m. We now want to ensure that no negative drawdown occurs anywhere at 100 m. At the same time, we still require a drawdown smaller than -2 m at the center of the excavation.

Mathematically, we state this optimization problem as follows:

> objective function:
> $$\min Q$$
> subject to:
> $$Q \ge 0$$
> at the center of the excavation center:
> $$u_1 \cdot Q \le -2$$
> at 100 m:
> $$-u_2 \cdot Q \le 0$$

Implementing and solving the problem using the `linprog` function:

```python
# coordatines of target points
x_center, y_center = 0, 0
x_100, y_100       = 0, 100

# head change constraints
s_center = -2
s_100 = 0

# head changes corresponding to unit pumping
u_center = u_model.h(x_center, y_center)
u_100 = u_model.h(x_100, y_100)

# linear programming
c = [1]
A = [[u_center], [-u_100]]
b = [s_center, s_100]
linprog(c, A_ub=A, b_ub=b, bounds=(0, None))
```

The solver reports that the optimization problem is **infeasible**, which means that there is no possible solution that satisfies all the constraints simultaneously. In other words, the restrictions we imposed contradict each other, so there is no combination of pumping and injection rates that can meet every requirement at once. Because of this, the solver cannot compute a feasible solution and therefore stops immediately.

### Optimizing two pumping rates

To make the problem feasible, we will decouple the NS‑ and EW‑oriented pumping and injection wells. In this **new configuration**, the NS‑oriented pumping wells all operate at the same pumping rate, and the water they extract is reinjected through the NS‑oriented injection wells. The same principle applies to the EW‑oriented wells.

```python
# Pumping wells

# Top and bottom rows (EW-oriented wells)
pumping_wells_EW = []
for x in np.arange(-25, 30, 10):
    pumping_wells_EW.append((x, 25))
    pumping_wells_EW.append((x, -25))

# Left and right columns (NS-oriented wells)
pumping_wells_NS = []
for y in np.arange(-15, 20, 10):
    pumping_wells_NS.append((-35, y))
    pumping_wells_NS.append((35, y))
```

```python
# Injection wells

# Top and bottom rows (EW-oriented wells)
injection_wells_EW = []
for x in [-25, 0, 25]:
    injection_wells_EW.append((x, 40))
    injection_wells_EW.append((x, -40))

# Left and right columns (NS-oriented wells)
injection_wells_NS = []
for y in [-15, 0, 15]:
    injection_wells_NS.append((-50, y))
    injection_wells_NS.append((50, y))
```

As an example, we simulate the system in which the EW-oriented pumping wells extract at 300 m³/d and the NS-oriented pumping wells at 420 m³/d:

```python
Q_pump_EW = 300  # m³/d
Q_pump_NS = 420  # m³/d

Q_inj_EW = -len(pumping_wells_EW) * Q_pump_EW / len(injection_wells_EW)
Q_inj_NS = -len(pumping_wells_NS) * Q_pump_NS / len(injection_wells_NS)

print('EW injection rate:', Q_inj_EW, 'm³/d')
print('NS injection rate:', Q_inj_NS, 'm³/d')
```

Building the model:

```python
model = Model(T)
for pumping_well in pumping_wells_EW:
    model.add_well(Q_pump_EW, *pumping_well)
for pumping_well in pumping_wells_NS:
    model.add_well(Q_pump_NS, *pumping_well)
for injection_well in injection_wells_EW:
    model.add_well(Q_inj_EW, *injection_well)
for injection_well in injection_wells_NS:
    model.add_well(Q_inj_NS, *injection_well)
```

Verifying that the sum of all pumping and injection rates is zero:

```python
sum([well.Q for well in model.wells])
```

Plotting the head change cross-sections:

```python
cs = np.linspace(-100, 100, 1000)
z = np.zeros(len(cs))
plt.figure(figsize=(12, 4))
plt.plot(cs, model.h(cs, z), "b-", label="east-west")
plt.plot(cs, model.h(z, cs), "k-", label="north-south")
plt.legend()
plt.grid()
plt.xlabel("distance (m)")
plt.ylabel("head change (m)");
```

We still observe a small drawdown at 100 m. So, let's apply linear programming to solve the problem!

The problem statement is as follows:

> objective function:
> $$\min \sum_i Q_i$$
> subject to:
> $$Q_i \ge 0$$
> at the excavation:
> $$\sum_i u_{1,i} Q_i \le -2$$
> at 100 m:
> $$-\sum_i u_{2,i} Q_i \le 0$$

where $i=1$ corresponds to the EW-oriented wells and $i=2$ to the NS-oriented wells.

So, we need **two models** to solve this problem:
1. the first model simulates the EW-oriented pumping wells extracting at unit rate and the corresponding EW-oriented injection wells reinjecting the extracted water;
2. the second model simulates the NS-oriented pumping wells extracting at unit rate and the corresponding NS-oriented injection wells reinjecting the extracted water.

```python
# model 1: EW-oriented wells
Q_pump_EW = 1
Q_inj_EW = -len(pumping_wells_EW) * Q_pump_EW / len(injection_wells_EW)
model_EW = Model(T)
for pumping_well in pumping_wells_EW:
    model_EW.add_well(Q_pump_EW, *pumping_well)
for injection_well in injection_wells_EW:
    model_EW.add_well(Q_inj_EW, *injection_well)

# model 2: NS-oriented wells
Q_pump_NS = 1
Q_inj_NS = -len(pumping_wells_NS) * Q_pump_NS / len(injection_wells_NS)
model_NS = Model(T)
for pumping_well in pumping_wells_NS:
    model_NS.add_well(Q_pump_NS, *pumping_well)
for injection_well in injection_wells_NS:
    model_NS.add_well(Q_inj_NS, *injection_well)
```

Defining the constraints:

```python
# at the center of the excavation
xy_center = (0, 0)
s_center = -2

# at 100 m
xy1_100 = (100, 0)
xy2_100 = (0, 100)
s_100 = 0
```

Note that, for safety, the ‘no‑drawdown’ constraint is enforced at both (0, 100) and (100, 0)!

Solving the problem using function `linprog`:

```python
c = [1, 1]
A = [[model_EW.h(*xy_center), model_NS.h(*xy_center)],
     [-model_EW.h(*xy1_100), -model_NS.h(*xy1_100)],
     [-model_EW.h(*xy2_100), -model_NS.h(*xy2_100)]]
b = [s_center, s_100, s_100]

res = linprog(c, A_ub=A, b_ub=b, bounds=[(0, None), (0, None)])
if res.success:
    Q_opt = res.x
    print(res.message)
    print("Q EW:", Q_opt[0], "m³/d")
    print("Q NS:", Q_opt[1], "m³/d")
else:
    print(res.message)
```

Let's build a model with the optimal pumping rates:

```python
Q_pump_EW = Q_opt[0]
Q_pump_NS = Q_opt[1]
Q_inj_EW = -len(pumping_wells_EW) * Q_pump_EW / len(injection_wells_EW)
Q_inj_NS = -len(pumping_wells_NS) * Q_pump_NS / len(injection_wells_NS)

model_opt = Model(T)
for pumping_well in pumping_wells_EW:
    model_opt.add_well(Q_pump_EW, *pumping_well)
for pumping_well in pumping_wells_NS:
    model_opt.add_well(Q_pump_NS, *pumping_well)
for injection_well in injection_wells_EW:
    model_opt.add_well(Q_inj_EW, *injection_well)
for injection_well in injection_wells_NS:
    model_opt.add_well(Q_inj_NS, *injection_well)
```

Plotting the cross-sections again to verify the result:

```python
cs = np.linspace(-100, 100, 1000)
z = np.zeros(len(cs))
plt.figure(figsize=(12, 4))
plt.plot(cs, model_opt.h(cs, z), "b-", label="east-west")
plt.plot(cs, model_opt.h(z, cs), "k-", label="north-south")
plt.legend()
plt.grid()
plt.xlabel("distance (m)")
plt.ylabel("head change (m)");
```

We clearly see that there is no negative drawdown anymore at 100 m from the center of the excavation, and that the drawdown at the center is -2 m.

### Adding a third constraint

However, we do not want a head change greater than 2 m at the injection wells, so we add a constraint for these wells:

> $$\sum_i u_{3,i} Q_i \le 2$$


We define the coordinates of the injection wells and the corresponding head change constraint:

```python
xy1_inj = (0, 40)
xy2_inj = (50, 0)
s_inj = 2
```

And we solve the new problem:

```python
c = [1, 1]
A = [[model_EW.h(*xy_center), model_NS.h(*xy_center)],
     [-model_EW.h(*xy1_100), -model_NS.h(*xy1_100)],
     [-model_EW.h(*xy2_100), -model_NS.h(*xy2_100)],
     [model_EW.h(*xy1_inj), model_NS.h(*xy1_inj)],
     [model_EW.h(*xy2_inj), model_NS.h(*xy2_inj)]]
b = [s_center, s_100, s_100, s_inj, s_inj]

res = linprog(c, A_ub=A, b_ub=b, bounds=[(0, None), (0, None)])
if res.success:
    Q_opt = res.x
    print(res.message)
    print("Q EW:", Q_opt[0], "m³/d")
    print("Q NS:", Q_opt[1], "m³/d")
else:
    print(res.message)
```

Building a model using these optimal pumping and injection rates:

```python
Q_pump_EW = Q_opt[0]
Q_pump_NS = Q_opt[1]
Q_inj_EW = -len(pumping_wells_EW) * Q_pump_EW / len(injection_wells_EW)
Q_inj_NS = -len(pumping_wells_NS) * Q_pump_NS / len(injection_wells_NS)

model_opt = Model(T)
for pumping_well in pumping_wells_EW:
    model_opt.add_well(Q_pump_EW, *pumping_well)
for pumping_well in pumping_wells_NS:
    model_opt.add_well(Q_pump_NS, *pumping_well)
for injection_well in injection_wells_EW:
    model_opt.add_well(Q_inj_EW, *injection_well)
for injection_well in injection_wells_NS:
    model_opt.add_well(Q_inj_NS, *injection_well)
```

Verifying that the head change constraints are satisfied:

```python
for xy in [xy_center, xy1_100, xy2_100, xy1_inj, xy2_inj]:
    print(f'head change at {xy}:', model_opt.h(*xy), 'm')
```

Let's redraw the cross-sectional plot:

```python
cs = np.linspace(-100, 100, 1000)
z = np.zeros(len(cs))
plt.figure(figsize=(12, 4))
plt.plot(cs, model_opt.h(cs, z), "b-", label="east-west")
plt.plot(cs, model_opt.h(z, cs), "k-", label="north-south")
plt.legend()
plt.grid()
plt.xlabel("distance (m)")
plt.ylabel("head change (m)");
```

We see that there is no drawdown at 100 m from the center of the excavation, that the drawdown at the center is -2 m, and that the head rise at the injection wells is not greater than 2m.
