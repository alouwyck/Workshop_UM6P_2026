# **Part 6 - The Principle of Superposition**

## Introduction

Groundwater flow in many practical situations can be approximated as a **linear system**, allowing the use of analytical solutions and the **principle of superposition**. Under steady-state conditions in a homogeneous aquifer that behaves approximately linearly, the **Thiem equation** provides a linear relationship between well discharge and hydraulic head. By combining multiple wells—both pumping and injection—the total hydraulic head distribution can be computed as the sum of individual contributions. This is a typical application of **spatial superposition** in groundwater flow modeling.

This approach can be used in practice for **groundwater management**. Typical applications include construction dewatering, where groundwater levels are temporarily lowered, as well as artificial recharge or reinjection, where extracted groundwater is returned to the aquifer to maintain balance. **Reinjection** is commonly applied to reduce drawdown outside the area of interest, limit environmental impacts, and ensure a sustainable use of groundwater resources.

In this notebook, we use superposition to simulate systems with **multiple pumping and reinjection wells**. This provides a fast and intuitive framework to explore how well configurations and discharge rates influence the hydraulic head distribution, and forms the basis for subsequent optimization of groundwater management strategies.

In addition to spatial superposition, we also make use of **superposition in time**. Many analytical solutions for transient groundwater flow, such as the classical **Theis solution**, express drawdown as a linear function of pumping rate. This linearity implies that time‑varying discharge rates can be simulated by decomposing them into a sequence of constant‑rate steps, each contributing its own transient response. By summing these individual responses, we can efficiently model **aquifer tests with variable pumping schedules**, intermittent pumping, or step‑drawdown experiments.

This time‑domain superposition provides a flexible and computationally light framework to explore how temporal changes in discharge affect the evolution of hydraulic heads, and complements the spatial superposition approach used for systems with multiple wells. In this notebook, we explore two classic examples: simulating a recovery test and modeling a pumping test near a river.

For a more detailed discussion of the principle of superposition and its applications, see [Louwyck (2023)](https://github.com/alouwyck/PhD).

## The Thiem Equation

As outlined in the first notebook of this workshop, the [Thiem equation](https://en.wikipedia.org/wiki/Aquifer_test#Steady-state_Thiem_solution) provides the analytical solution to the one‑dimensional, axisymmetric, steady‑state groundwater‑flow equation describing radial flow to a wel:

> $h(r) = h(R) + \dfrac{Q}{2 \pi T} \ln \dfrac{r}{R}$

where

- $h$ is the hydraulic head [L]
- $r$ is the radial distance [L] to the pumping well
- $R$ is the radius of influence [L] of the extraction
- $Q$ is the pumping rate [L³/T], which is positive in case of extraction
- $T$ is the aquifer transmissivity [L²/T]

The drawdown $s$ [L] is defined as:

> $s(r) = h(r) - h(R)$

with $h(R)$ the initial head before pumping. Note that the drawdown is positive in case of extraction.

Combining the two equations gives:

> $s(r) = \dfrac{Q}{2 \pi T} \ln \dfrac{r}{R}$

Drawdown is negative when the hydraulic head decreases, and positive when it increases. Drawdown is zero at distance $R$; hence, the term radius of influence.



Let's define a function `thiem` that implements this formula:

```python
import numpy as np

def thiem(Q, T, R, r):
    '''
    Simulate drawdown using the Thiem equation
    Q: pumping rate [L³/T] - float
    T: aquifer transmissivity [L²/T] - float
    R: radius of influence [L] - float
    r: radial distance [L] - array-like
    returns drawdown s [L] for each distance r
    '''
    return Q/2/np.pi/T * np.log(np.array(r)/R)
```

Testing the function for a pumping well extracting groundwater at a rate of 100 m³/d from a homogeneous aquifer with transmissivity of 50 m²/d and radius of influence equal to 500 m:

```python
Q = 100  # pumping rate (m³/d)
T = 50   # transmissivity (m²/d)
R = 500  # radius of influence (m)

r = np.logspace(-1, np.log10(R), 100)  # radial distances (m)

s = thiem(Q, T, R, r)  # drawdown (m)
```

Plotting the result:

```python
import matplotlib.pyplot as plt

plt.semilogx(r, s)
plt.xlabel('$r$ (m)')
plt.ylabel('$s$ (m)')
plt.title('Thiem equation')
plt.grid();
```

We observe a linear relationship between drawdown $s$ and the logarithm of distance $r$. There is also a linear relationship between drawdown $s$ and pumping rate $Q$:

```python
r = 1  # radial distance (m)
Qs = range(100, 1010, 10)  # range of pumping rates (m³/d)

s = [thiem(Q, T, R, r) for Q in Qs]  # drawdown (m)

plt.plot(Qs, s)
plt.xlabel('$Q$ (m³/d)')
plt.ylabel('$s$ (m)')
plt.title(f'$r$ = {r} m')
plt.grid();
```

It is precisely this linear dependence of drawdown on pumping rate that enables the use of the principle of superposition.

## Spatial Superposition

A key property of a homogeneous aquifer that behaves linearly is the **[principle of superposition](https://en.wikipedia.org/wiki/Superposition_principle)**, which states that the combined effect of multiple stresses is equal to the sum of the effects of each stress applied separately.

For a single well, the Thiem equation gives the hydraulic head $h$ as a function of radial distance $r$ to the well. Because this relationship is linear in the pumping rate $Q$, the contributions of multiple wells can simply be added.

In practice, this means that for a system with several pumping and injection wells:

* Each well creates its own cone of depression (pumping) or mounding (injection).
* The total hydraulic head $h$ at any point $(x,y)$ is obtained by summing the contributions of all wells.
* Pumping wells (positive $Q$) lower the head $h$, while injection wells (negative $Q$) increase it.

An important implication is that complex well fields can be analyzed by **combining simpler solutions**. This makes superposition a powerful and efficient tool for simulating groundwater systems, especially when exploring different well configurations or evaluating the effect of reinjection strategies.

Mathematically, the total hydraulic head $h$ in any point $(x,y)$ is computed using the **Thiem equation**:

> $h(x,y) = h_0 + \sum\limits_{i=1}^{n} \dfrac{Q_i}{2\pi T} \ln \dfrac{r_i}{R_i}$

with $h_0$ the initial head [L] in the aquifer before pumping, and $n$ the number of wells. Each term in the summation represents the influence of a single well $i$ at location $(x_{w,i}, y_{w,i})$, and $r_i$ is the Euclidean distance [L] to the evaluation point $(x, y)$:

> $r_i = \sqrt{\left(x-x_{w,i}\right)^2 + \left(y-y_{w,i}\right)^2}$



A closer examination of the superposition formula reveals a conceptual issue with the radius of influence $R_i$, which only has a clear physical meaning for a single pumping well, where the problem is axially symmetric. In the presence of multiple wells, this assumption of axial symmetry no longer holds, and $R_i$ must instead be interpreted as the distance to a point $(x_0, y_0)$ where the hydraulic head remains constant and equal to the initial head $h_0$:

> $R_i = \sqrt{\left(x_0-x_{w,i}\right)^2 + \left(y_0-y_{w,i}\right)^2}$

Essentially, this corresponds to defining a **constant‑head or [Dirichlet boundary condition](https://en.wikipedia.org/wiki/Dirichlet_boundary_condition)** at the point $(x_0, y_0)$.

If we place this constant‑head point at a sufficiently large distance from the wells, we may assume that for all wells $i$:

> $R_i \approx R$.

In that case, the superposition equation can be reformulated as follows:

> $h(x,y) = h_0 + \dfrac{1}{2\pi T} \sum\limits_{i=1}^{n} Q_i \ln r_i - \dfrac{\ln R}{2\pi T} \sum\limits_{i=1}^{n} Q_i$

Finally, if we impose that the **sum of all pumping and injection rates equals zero**:

> $\sum\limits_{i=1}^{n} Q_i = 0$

then the term containing $R$ drops out:

> $h(x,y) = h_0 + \dfrac{1}{2\pi T} \sum\limits_{i=1}^{n} Q_i \ln r_i$

Placing the constant‑head boundary condition far away from the wells is considered best practice in groundwater‑flow modeling, as it prevents this boundary from influencing the flow toward the wells. The requirement that the sum of all pumping rates must equal zero effectively corresponds to reinjecting the abstracted water, which is also good practice in dewatering operations. Hydrodynamically, it implies that the system will eventually reach a new steady state, making the application of the Thiem equation appropriate in this context.

Let's demonstrate this with a simple example of a pumping well and an injection well located 50 meters apart in a homogeneous aquifer:

```python
T = 100  # aquifer transmissivity
Q = 300  # pumping rate

# well coordatines
x_pump, y_pump = 25, 0  # pumping well
x_inj, y_inj = -25, 0   # injection well

# grid
x, y = np.meshgrid(np.linspace(-100, 100, 100),  # x-axis: 100 point between -100 and 100
                   np.linspace(-50, 50, 100))    # y-axis: 100 point between -50 and 50

# calculate distances to wells
r_pump = np.sqrt((x-x_pump)**2 + (y-y_pump)**2)
r_inj = np.sqrt((x-x_inj)**2 + (y-y_inj)**2)

# head change
s = (Q*np.log(r_pump) - Q*np.log(r_inj)) / 2/np.pi/T
```

In this example, the aquifer transmissivity is 100 m²/d and the pumping rate $Q$ is 300 m³/d, which implies an injection rate of $-Q$, or –300 m³/d. The pumping well is located at $(25, 0)$ and the injection well at $(-25, 0)$, giving a separation distance of 50 m. A grid with 100 by 100 nodes $(x, y)$ is constructed, and for each grid node the Euclidean distance $r$ to both wells is computed. The head change $s$ is obtained by setting the initial head $h_0$ to zero:

> $s(x,y) = \dfrac{\sum\limits_{i=1}^{n} Q_i \ln r_i}{2\pi T}$

The resulting contour plot shows a clear drawdown cone around the pumping well and a recharge mound around the injection well. Between the two wells, a distinct groundwater divide appears: a vertical line along which the hydraulic head does not change. This line can therefore be interpreted as a constant‑head boundary. In fact, this illustrates how the effect of a pumping well near a river or canal with a fixed water level can be simulated in a simple way — by adding an imaginary injection well. This is the essence of the so‑called **[method of images](https://en.wikipedia.org/wiki/Method_of_images)**.

```python
plt.figure(figsize=(10, 5))
plt.contourf(x, y, s)
plt.colorbar()
ax = plt.gca() # Get the current axes
ax.set_aspect('equal') # Set aspect ratio to equal
```

Let's verify whether superimposing the Thiem equation with a large $R$ yields the same result:

```python
s_pump = thiem(Q, T, 1E5, r_pump)  # pumping
s_inj = thiem(-Q, T, 1E5, r_inj)   # injection: -Q!!
s_total = s_pump + s_inj
print(np.all(np.isclose(s_total, s)))
```

We get the same result indeed!

## Using OO Python

To make the code more structured, reusable, and aligned with how groundwater systems are conceptually organized, we now switch to an object‑oriented (OO) design: wells become objects that belong to a groundwater-flow model object, creating a clear many‑to‑one relation that reflects the real modeling.

So the code below applies OO in Python to model groundwater flow as a set of interacting components. The `Model` class represents the groundwater-flow model, while each `Well` object represents an individual pumping or injection well. This separation allows each concept to have its own clear responsibilities.

The `Well` class encapsulates everything related to a single well: its location `(xw,yw)`, pumping rate `Q`, well-radius `rw`, and the `model` to which it belongs. The method `s(x,y)` computes and returns the drawdown caused by that well only. Because each well is an independent object, it becomes easy to add, remove, or inspect wells without affecting the rest of the model.

```python
class Well:
    """
    A pumping or injection well represented by the Thiem equation.

    Parameters
    ----------
    model : Model
        The Model object to which the well belongs.
    Q : float
        Pumping rate [L³/T]. Positive for pumping, negative for injection.
    xw, yw : float
        Coordinates [L] of the well.
    rw : float, optional
        Well radius [L]. Distances smaller than this are set to rw
        to avoid singularities. Default is 0.1 m.
    """

    def __init__(self, model, Q, xw, yw, rw=0.1):
        self.model = model
        self.Q = Q
        self.xw, self.yw = xw, yw
        self.rw = rw

    def s(self, x, y):
        """
        Compute drawdown at coordinates (x, y) due to this well
        using the Thiem equation.

        Parameters
        ----------
        x, y : float or array-like
            Coordinates [L] at which the drawdown is evaluated.

        Returns
        -------
        ndarray or float
            Drawdown value(s) [L] at the given coordinate(s).
        """
        x, y = np.array(x), np.array(y)
        r = np.sqrt((x - self.xw)**2 + (y - self.yw)**2)
        r[r < self.rw] = self.rw  # prevent singularity

        return thiem(self.Q, self.model.T, self.model.R, r)
```

The `Model` class manages the shared properties: aquifer transmissivity `T`, initial head `h0`, and approximate distance `R` of each well to the constant-head point. It also maintains a list of `well` objects. When `add_well()` is called, a new `Well` object is created and stored inside the model. The method `h(x,y)` then computes the total hydraulic head by looping over all wells and combining their individual contributions through superposition. In this sense, the model acts as a coordinating object that aggregates the behaviour of multiple wells.

```python
class Model:
    """
    Model for single homogeneous aquifer represented by transmissivity, initial head,
    and a list of wells.

    Parameters
    ----------
    T : float
        Aquifer transmissivity [L²/T].
    h0 : float, optional
        Initial hydraulic head [L] in the aquifer. Default is 0.
    R : float, optional
        Approximate distance [L] to constant-head point for wells in the aquifer.
        Default is 1e6.
    """

    def __init__(self, T, h0=0, R=1e6):
        self.T = T
        self.h0 = h0
        self.R = R
        self.wells = []

    def add_well(self, Q, xw, yw, rw=0.1):
        """
        Add a pumping or injection well to the model.

        Parameters
        ----------
        Q : float
            Pumping rate [L³/T].
        xw, yw : float
            Well coordinates [L].
        rw : float, optional, default is 0.1
            Well radius [L].

        Returns
        -------
        Well
            The newly created Well object.
        """
        well = Well(self, Q, xw, yw, rw)
        self.wells.append(well)
        return well

    def h(self, x, y):
        """
        Compute the hydraulic head at coordinates (x, y),
        obtained by superposition of all wells.

        Parameters
        ----------
        x, y : float or array-like
            Coordinates [L] at which the head is evaluated.

        Returns
        -------
        ndarray or float
            Hydraulic head [L] value(s) at the given coordinate(s).
        """
        return self.h0 + np.sum([well.s(x, y) for well in self.wells], axis=0)
```

This OO structure keeps the logic modular:

- each well knows how to compute its own effect;
- the model knows how the system behaves as a whole;
- the user interacts with the model at a high level by creating objects rather than manually managing equations.

This makes the code easier to extend (e.g., adding new well types or boundary conditions) and clearer to understand in a groundwater‑modeling context.

Let's re-implement the example of combined pumping and injection using the two defined classes:

```python
model = Model(T)                          # model
pump = model.add_well(Q, x_pump, y_pump)  # pumping well
inj = model.add_well(-Q, x_inj, y_inj)    # injection well

# compare drawdowns
print(np.all(np.isclose(s, model.h(x, y))))
```

## TimML

The Python library [TimML](https://timml.readthedocs.io/en/latest/) developed by Prof. [Mark Bakker](https://www.tudelft.nl/citg/over-faculteit/afdelingen/watermanagement/medewerker/staff-hydrology/academic-staff/profdrir-mark-bakker) also applies the principle of superposition to solve complex flow problems by combining simpler analytical elements. This is why the approach is known as the [Analytic Element Method](https://en.wikipedia.org/wiki/Analytic_element_method) (AEM). It also uses an object-oriented design that is similar to the two classes we defined in previous section.

First, we need to install the TimML package:

```python
!pip install timml -q
```

We import `timml` and create a `model` object, which accepts the aquifer properties:

- hydraulic conductivity `kaq` is set to transmissivity `T`
- top and bottom `z` of the aquifer are set to 1 and 0 to obtain an aquifer thickness of 1



```python
import timml

model_tim = timml.ModelMaq(
    kaq=T,      # hydraulic conductivity (m/d) is set to transmissivity T
    z=[1, 0],   # top and bottom of aquifer (m) is set to 1 and 0 to get an aquifer thickness of 1
)
```

Note that the aquifer thickness $D$ is set to 1, so that the hydraulic conductivity $K$ of the aquifer is equal its transmissivity $T$:

> $T = KD$


We add the pumping well:

```python
pump_tim = timml.Well(
    model=model_tim,  # model to which the element is added
    xw=pump.xw,       # x-location of well (m)
    yw=pump.yw,       # y-location of well (m)
    Qw=pump.Q,        # discharge of well (m³/d), positive for extraction
    rw=pump.rw,       # well radius (m)
)
```

and the injection well:

```python
inj_tim = timml.Well(
    model=model_tim,  # model to which the element is added
    xw=inj.xw,        # x-location of well (m)
    yw=inj.yw,        # y-location of well (m)
    Qw=inj.Q,         # discharge of well (m³/d), negative for injection
    rw=inj.rw,        # well radius (m)
)
```

We also have to define a `Constant` head at distance `R`:

```python
constant = timml.Constant(
    model=model_tim,   # model to which the element is added
    xr=model.R,        # x-location of fixed head (m)
    yr=0,              # y-location of fixed head (m)
    hr=0,              # fixed head (m) which is set to zero to obtain drawdowns
)
```

Once we've added all the elements, we are ready to `solve()` the model:

```python
model_tim.solve()
```

TimML provides a method `headgrid2()` to compute the heads on a given meshgrid:

```python
s_tim = model_tim.headgrid2(-100, 100, 100,
                            -50, 50, 100)

# compare drawdowns
s_tim = s_tim[0]  # first layer
print(np.all(np.isclose(s_tim, s, atol=1e-4)))
```

The TimML results match ours up to four decimal places.

TimML is designed to simulate groundwater flow in multilayer aquifer systems, which is why we need to extract the hydraulic heads from the first layer `s_tim[0]`. Because we set the constant head to zero, the model actually returns head changes instead of heads.

## Simulation of Aquifer Reinjection

### Setup of the construction dewatering system

The example consists of a rectangular excavation of 50 m × 30 m, centered in the model domain. To enable dry working conditions, groundwater levels inside the excavation are lowered using a series of pumping wells placed around the site.

The pumping wells are arranged in a rectangular pattern at a distance of 10 m from the excavation boundary, whereas the distance between two adjacent wells is also 10 m. This creates a controlled drawdown that keeps the groundwater level below the excavation base.

To reduce the impact of dewatering on the surrounding area, the extracted groundwater is re-injected into the aquifer using a second ring of wells located further away. These injection wells are positioned at a distance of 25 m from the excavation boundary, forming a wider perimeter around the site.

The purpose of the reinjection is to compensate for the extracted water, thereby limiting drawdown outside the construction area and maintaining the regional groundwater balance.

### Simulation of the system using the superposition principle

Now we will model the system using the principle of superposition applied to the Thiem equation as explained in the previous sections. Recall that each well — whether pumping or injection — contributes to the hydraulic head distribution, and the total effect is obtained by summing all contributions.

We consider a homogeneous aquifer with constant thickness equal to 20 m and horizontal conductivity equal to 10 m/d:

```python
T = 20 * 10  # transmissivity T = KD
```

There are 20 pumping wells surrounding the excavation, where each well extracts groundwater at a rate $Q_{\text{pump}}$ of 240 m³/d. The injection wells reinject the extracted water at a constant rate ${Q_{\text{inj}}}$ equal to:

> $Q_{\text{inj}} = -\dfrac{n_{\text{pump}} \cdot Q_{\text{pump}}}{n_{\text{inj}}}$

with $n_{\text{pump}}$ and $n_{\text{inj}}$ the number of pumping and injection wells, respectively.

```python
n_pump, n_inj = len(pumping_wells), len(injection_wells)

Q_pump = 240
Q_inj = -n_pump * Q_pump / n_inj

print('pumping rate:', Q_pump, 'm³/d')
print('injection rate:', Q_inj, 'm³/d')
```

Now we're ready to define the superposition model and add the wells:

```python
model = Model(T)
for pumping_well in pumping_wells:
    model.add_well(Q_pump, *pumping_well)
for injection_well in injection_wells:
    model.add_well(Q_inj, *injection_well)
```

We compute the head change `s` for a 100 m x 100 m grid:

```python
x, y = np.meshgrid(np.linspace(-100, 100, 100),
                   np.linspace(-100, 100, 100))
s = model.h(x, y)
```

And we visualize the result with a contour plot:

```python
plt.figure(figsize=(10, 5))
levels = np.arange(-2.4, 1.4, 0.1)
plt.contourf(x, y, s, levels)
plt.colorbar();
```

Although we clearly observe the excavation and the individual wells, we also create a contour plot to which the exact positions of the excavation and the wells are added:

```python
ax = plot_wells()
levels = np.arange(-2.5, 1.51, 0.25)
cs = ax.contour(x, y, s, levels);
target_levels = [-1.5, -1.0, -0.5, 0.0]
selected_levels = [cs.levels[np.argmin(abs(cs.levels - tlevel))] for tlevel in target_levels]
ax.clabel(cs, levels=selected_levels, inline=True, fmt='%1.1f');
```

We also create north–south and east–west cross‑sections of the head change along the x‑ and y‑axes:

```python
cs = np.linspace(-100, 100, 1000)
z = np.zeros(len(cs))
plt.figure(figsize=(12, 4))
plt.plot(cs, model.h(cs, z), "b-", label="east-west")
plt.plot(cs, model.h(z, cs), "k-", label="north-south")
plt.legend()
plt.grid()
plt.xlabel("distance (m)")
plt.ylabel("head change (m)");
```

Finally, we compute the drawdown at the center and at the corners of the excavation:

```python
print('excavation center:', model.h([0], [0]).item(), 'm')
print('excavation corner:', model.h([excavation['xmin']], [excavation['ymin']]).item(), 'm')
```

### Simulation of the system using TimML

Using TimML, we first define the model, then add the wells and the constant‑head boundary, and finally solve the model:

```python
# define model
model_tim = timml.ModelMaq(
    kaq=T,
    z=[1, 0],
)

# add wells
for well in model.wells:
    timml.Well(model=model_tim,
        xw=well.xw, yw=well.yw,
        Qw=well.Q, rw=well.rw)

# add constant head
timml.Constant(model=model_tim, xr=model.R, yr=0, hr=0)

# solve model
model_tim.solve()
```

The contour plot is virtually identical, as expected!

```python
s_tim = model_tim.headgrid2(-100, 100, 100,
                            -100, 100, 100)[0]
plt.figure(figsize=(10, 5))
levels = np.arange(-2.5, 1.4, 0.1)
plt.contourf(x, y, s_tim, levels)
plt.colorbar();
```

And we obtain the same drawdowns (matching up to eight decimal places!) at the center and at the corners of the excavation:

```python
print('excavation center:', model_tim.head(0, 0).item(), 'm')
print('excavation corner:', model_tim.head(excavation['xmin'], excavation['ymin']).item(), 'm')
```

This simplified setup illustrates how well placement and discharge rates influence groundwater levels. It also provides a foundation for exploring more advanced questions, such as optimizing well rates or configurations to achieve efficient and environmentally responsible groundwater management.

### Optimizing the pumping rate

If we want to impose a head-change target (or constraint) at only a single point, the linearity of the model allows us to use simple **proportional scaling**: the entire solution can be rescaled directly.

For example, suppose we want to obtain a drawdown of -2 m at the center of the excavation. Because the model is linear, we can simply scale the result proportionally:

```python
s_target = -2
s_center = model.h([0], [0]).item()

Q_pump_opt = Q_pump / s_center * s_target
Q_inj_opt = -n_pump * Q_pump_opt / n_inj

print('optimal pumping rate:', Q_pump_opt)
print('optimal injection rate:', Q_inj_opt)
```

Redefining the model using these optimal rates:

```python
model_opt = Model(T)
for pumping_well in pumping_wells:
    model_opt.add_well(Q_pump_opt, *pumping_well)
for injection_well in injection_wells:
    model_opt.add_well(Q_inj_opt, *injection_well)
```

Replotting the cross-sections:

```python
cs = np.linspace(-100, 100, 1000)
z = np.zeros(len(cs))
plt.figure(figsize=(12, 4))
plt.plot(cs, model_opt.h(cs, z), "b-", label="east-west")
plt.plot(cs, model_opt.h(z, cs), "k-", label="north-south")
plt.legend()
plt.grid()
plt.xlabel("distance (m)")
plt.ylabel("head change (m)");
```

The drawdown at the center is indeed -2 m! Let’s perform one final verification:

```python
print('excavation center:', model_opt.h([0], [0]).item(), 'm')
print('excavation corner:', model_opt.h([excavation['xmin']], [excavation['ymin']]).item(), 'm')
```

## Superposition in Time

In this section, two well-known examples of the application of the superposition principle are given. The first example applies *superposition in time* to simulate a recovery test, while the second example applies the *method of images* to simulate the transient effect of a pumping well close to a river.

### Theis' recovery test

It is straightforward to simulate a recovery test in a confined aquifer by applying the superposition method to the Theis solution.

First, we import the necessary Python packages, and we clone the [`maxsypy`](https://github.com/alouwyck/maxsypy) package which implements several analytical and semi-analytical solutions for simulating axisymmetric groundwater flow discussed in [Louwyck (2023)](https://github.com/alouwyck/PhD):

```python
import numpy as np
import matplotlib.pyplot as plt
import inspect
import os, shutil

# remove directory 'maxsypy' first if it already exists
path = "maxsypy"
if os.path.exists(path):
    shutil.rmtree(path)

# clone maxsypy repository from Github
!git clone https://github.com/alouwyck/maxsypy

# copy data files to current working directory
data_folder = './data'
if os.path.exists(data_folder):
    shutil.rmtree(data_folder)
shutil.copytree("./maxsypy/seminaries/UM6P_jan_2025/data", data_folder)

# put maxsypy package directory in the current working directory
os.rename(path, 'temp')
shutil.copytree(os.path.join('temp', path), path);
shutil.rmtree('temp')
```

We also install and import `ttim`:

```python
!pip install ttim -q
import ttim
```

Now, we import function `theis_recovery` from module `maxsypy.special`. When we inspect the code, we see it uses function `theis` and it applies superposition in time. First, the pumping is simulated during the entire simulation period, that is during both the pumping and the recovery phase. Then the calculated drawdown for the recovery phase is corrected by superimposing the head change due to a well injecting water at the same rate.

```python
from maxsypy.special import theis_recovery

print(inspect.getsource(theis_recovery))
```

Let's illustrate this using the running example from the previous notebooks:

```python
Q  = -1000  # pumping rate (m³/d): Q < 0 -> extraction
T  = 200    # aquifer transmissivity (m²/d)
S  = 0.1    # aquifer storativity (-)
```

We simulate the head change in the pumping well with a radius of 0.1 m. The well extracts water from the confined aquifer during 1 day, after which its pump is shut down:

```python
r = 0.1    # radial distance (m)
t_end = 1  # time at which pump is shut down
t = np.logspace(-5, np.log10(t_end), 1000)
t = np.hstack((t, t_end + t))  # simulation times (d)
```

Using **TTim**, it is very straightforward to simulate this pumping and recovery test:

```python
model = ttim.ModelMaq(kaq=T, Saq=S, z=[1, 0],  # use T and S by setting aquifer thickness to unity
                      tmin=1e-5, tmax=10)

well = ttim.Well(model,
                 rw=1e-5,  # well-radius -> 0
                 tsandQ=[(0, Q),  # starting with the pumping phase
                         (1, 0)]  # recovery phase (Q=0) after 1 day
)

model.solve()

s_ttim = model.head(x=0, y=r, t=t).squeeze()
```

Let's compare the TTim simulation to the result obtained by applying the superposition method implemented in the `theis_recovery` function:

```python
for S2 in np.array([0.5, 1, 2]) * S:  # storativity (-) during recovery
    s = theis_recovery(r=r, t=t, Q=Q, T=T, S=S, t_end=t_end, S2=S2)  # drawdown (m)
    plt.plot(t, s.squeeze(), '-', label="S2 = " + str(S2));

plt.plot(t, s_ttim, 'k:');  # drawdown (m) simulated using TTim

plt.xlabel('t (d)');
plt.ylabel('s (m)');
plt.legend();
plt.grid();
```

Both approaches virtually give the same time-drawdown curve. It is possible the aquifer storativity `S2` during recovery is different from the storativity `S` during pumping. That is why additional simulations were performed where `S2` is two times smaller and two times larger than `S`. Although the factor 2 may seem unrealistically large, it is chosen to obtain a clear visual distinction.

### A pumping well close to a river

As already discussed in the previous section on spatial superposition, it is possible to simulate flow to a pumping well close to a straight river with constant water level by applying **the method of images**. Here, we illustrate the method by applying it to the Theis equation:

```python
from maxsypy.special import theis
```

We take the same example as before, and additionally, we define a north-south oriented river that is L = 100 m away from the pumping well.

```python
Q  = -1000  # pumping rate (m³/d): Q < 0 -> extraction
T  = 200    # aquifer transmissivity (m²/d)
S  = 0.1    # aquifer storativity (-)
L  = 100    # horizontal distance (m) between well and rivier
```

The well is at position (0, 0), and if the river is located on the left of the well, then we need to define an image well at position (-2L, 0) that injects water at the same rate -Q. Superimposing the head changes of both the pumping well and the imaginary injection well creates a constant-head boundary that coincides with the river:

```python
# applying the method of images

t = 1e3  # time (d)
x_pump, y_pump = 0, 0  # pumping well coordinates: (0, 0)
x_im, y_im = -2*L, 0   # image injection well coordinates: (-2L, 0)

x = y = np.linspace(-L, L, 100)     # x and y coordinates (m)
x_grid, y_grid = np.meshgrid(x, y)  # grid coordinates

r_pump = np.sqrt((x_grid - x_pump)**2 + (y_grid - y_pump)**2)  # distances (m) to pumping well
s_pump = theis(r=r_pump.flatten(), t=t, T=T, S=S, Q=Q);  # drawdown (m) due to pumping

r_im = np.sqrt((x_grid - x_im)**2 + (y_grid - y_im)**2)  # distances (m) to injection well
s_im = theis(r=r_im.flatten(), t=t, T=T, S=S, Q=-Q);  # head change (m) due to injection

s_theis = np.reshape(s_pump + s_im, x_grid.shape)
plt.contourf(x_grid, y_grid, s_theis);
plt.colorbar()
plt.gca().set_aspect('equal');
plt.xlabel('x (m)');
plt.ylabel('y (m)');
plt.title('drawdown s (m)', fontdict=dict(fontsize=10));
```

Recall that the radial distance $r$ between the well $(x_w, y_w)$ and a grid node $(x,y)$ is determined by the Euclidean distance between both points:

> $r = \sqrt{(x-x_w)^2 + (y-y_w)^2}$

The method of images can also be applied to simulate the effect of an impervious lateral boundary close to to the well. In this case, an image well extracting at the same pumping rate is used instead of an imaginary injection well.

As we already know, the Theis equation may be approximated by the Thiem equation for large values of time. The question that comes into mind here is which value for the radius of influence `R` do we have to choose? As explained in the previous section on spatial superposition, the value for `R` doesn't matter as long as it is sufficiently large when the pumping and the injection are balanced:

```python
from maxsypy.special import thiem

R = 1e6  # setting R to a very large value
s_pump = thiem(r=r_pump.flatten(), T=T, Q=Q, r_out=R);  # drawdown (m) due to pumping
s_im = thiem(r=r_im.flatten(), T=T, Q=-Q, r_out=R);  # head change (m) due to injection
s_thiem = np.reshape(s_pump + s_im, x_grid.shape)

plt.contourf(x_grid, y_grid, s_thiem);
plt.colorbar()
plt.gca().set_aspect('equal');
plt.xlabel('x (m)');
plt.ylabel('y (m)');
plt.title('drawdown s (m)', fontdict=dict(fontsize=10));
```

As expected, we get the same contour plot as before!

Of course, the method of images can also be applied using **TTim**. However, the package also provides line sinks, which can be used to simulate the effect of rivers of all shapes. Let's illustrate this by adding a `HeadLineSinkString` to the corresponding Theis model:

```python
# start with building the Theis model
model = ttim.ModelMaq(kaq=T, Saq=S, z=[1, 0], tmin=0.001, tmax=t)
well = ttim.Well(model, rw=1e-5, tsandQ=[(0, -Q)])

# add the line sink to the model
n = 50                             # the line sink is discretized into 50 nodes
ymax = 1000                        # maximum y-coordinate of the line sink
xls = -L * np.ones(n)              # x-coordinates of the nodes; the river is located at x=-L
yls = np.linspace(-ymax, ymax, n)  # y-coordinates of the nodes ranging from -ymax to ymax

river = ttim.HeadLineSinkString(
    model,                                 # add line sink to the model
    xy=list(zip(xls, yls, strict=False)),  # (x, y) coordinates of the nodes
    tsandh="fixed"                         # the river has a fixed level
)

# solve the model
model.solve()
```

Creating a contour plot with TTim takes too much time, and therefore, the head for a vertical cross-section is calculated:

```python
irow = 50  # cross-section corresponding to horizontal line y = y[irow]
s_ttim = np.array([model.head(x=x, y=y[irow], t=t) for x in x]).squeeze()
```

To see how the cone of depression expands over time, we also compute the drawdown curves along this cross‑section after 0.01, 0.1, 1, and 10 days using the method of images:

```python
t = np.logspace(-2, 1, 4)                                  # 0.01, 0.1, 1, 10 days
r_pump = np.sqrt((x - x_pump)**2 + (y[irow] - y_pump)**2)  # radial distances (m) to pumping well
r_im = np.sqrt((x - x_im)**2 + (y[irow] - y_im)**2)        # radial distances (m) to imaginary well
s_extra = theis(r=r_pump.flatten(), t=t, T=T, S=S, Q=Q) +\
          theis(r=r_im.flatten(), t=t, T=T, S=S, Q=-Q)     # superimposing head changes
```

Plotting the result:

```python
plt.plot(x, s_theis[irow, :], 'b-', label='Theis');
plt.plot(x, s_thiem[irow, :], 'y--', label='Thiem');
plt.plot(x, s_ttim, 'r:', label='TTim');
plt.plot(x, s_extra, color="gray", linestyle="--");
plt.legend(loc='lower left');
plt.grid();
plt.xlabel('x (m)');
plt.ylabel('s (m)');
plt.title(f'y = {np.round(y[irow])} m');
```

The three approaches virtually give the same results! When using TTim, the line sink should be long enough and have a sufficient number of nodes. The grey dashed curves clearly show how the cone of depression expands. After 1 day, it reaches the river, causing the cone to lose its axial symmetry.
