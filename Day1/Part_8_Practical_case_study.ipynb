## Drainage system of combined pumping and infiltration

In this section we simulate the effect of the drainage system at the excavation site 'Duinenabdij'. The system consists of pumping and injections wells. The total pumping rate equals the total injection rate as all the extracted groundwater is re-injected into the aquifer system. As a consequence, the system will finally reach a steady state. Therefore, we use class `Steady` to simulate the individual head change of each well. The total head change is obtained by summing all individual head changes according to the **superposition principle**.

```python
from maxsypy.multilayer import Steady
```

The aquifer system consists of three permeable layers separated by two semi-pervious layers. As we assume the flow is in a steady state, the latter can be conceptualized as incompressible zero-thickness resistance layers:

```python
# transmissivities T (m²/d)
T = [2.9 * 12, 5.2 * 13.8, 10.7 * 42.0]  # T = D * Kh

# resistances c (d)
c = [1300.0, 49.70]  # confined system
```

Note that only half of the injection wells recharge the lower layer.

The original pumping and injection rates (m³/d) are as follows:

- The discharge from the upper layer through the gravel pack surrounding the pumping wells is estimated at 4 m³/d.
- The pumping rate of the wells screened in the middle layer is 60 m³/d.
- The injection rate of the wells in the lower layer is –40 m³/d.
- The injection rate of the wells in the middle layer is determined by balancing the total injection rate with the total pumping rate.

```python
# pumping and injection rates
npw, niw = len(pwells), len(iwells)
print(f"{npw} pumping wells")
print(f"{niw} injection wells")
Q_grav = 4.0  # discharge through gravel pack surrounding the pumping wells (upper layer)
Q_pump = 60.0  # pumping rate (middle layer)
Q_deep = -40.0  # deep injection rate (lower layer)
Q_inj = -(npw * (Q_pump + Q_grav) + niw / 2 * Q_deep) / niw  # injection rate (middle layer)
print(f"gravel pack discharge (upper layer): {Q_grav:.2f} m³/d")
print(f"pumping rate (middle layer): {Q_pump:.2f} m³/d")
print(f"injection rate (middle layer): {Q_inj:.2f} m³/d")
print(f"injection rate (lower layer): {Q_deep:.2f} m³/d")
```

Creating the superposition grid:

```python
n = 100  # 100 by 100 grid nodes
coord = np.linspace(0, 600, n)
xgrid, ygrid = np.meshgrid(coord, coord)  # 600m x 600m meshgrid
```

Defining function `euclidean` to calculate Euclidean distance between two points:

```python
euclidean = lambda x1, y1, x2, y2: np.sqrt((x1 - x2)**2 + (y1 - y2)**2)
```

As the aquifer system is confined and steady flow is simulated, the system cannot be laterally unbounded. Therefore we need to define a point at a very large distance at which drawdown is zero. This is a constant-head boundary that will determine the outer model boundary of each axisymmetric model that will be used to simulate the head change due to the individual wells. Because extraction and injection are balanced and the point is at very large distance, this boundary condition will have no significant impact on the total head change.

```python
# constant-head point (x0, y0) at very large distance
x0, y0 = 1e5, 1e5
```

Finally, we apply the superposition method:

```python
def superposition(pwells, iwells, Q_grav, Q_pump, Q_inj, Q_deep, xgrid, ygrid):

    # initializing the head change array
    s = np.zeros((len(T), n, n))

    # calculate the head change caused by each pumping well
    for xy in pwells:
        x, y = tuple(xy)
        r_out = euclidean(x, y, x0, y0)  # determine the outer model boundary using (x0, y0)
        model = Steady(T=T, c=c, Q=[Q_grav, Q_pump, 0], r_out=r_out)
        r = euclidean(x, y, xgrid, ygrid)
        s += model.h(r=r.flatten()).reshape(s.shape)

    # calculate the head change caused by each injection well
    for i, xy in enumerate(iwells):
        x, y = tuple(xy)
        r_out = euclidean(x, y, x0, y0)
        Q3 = Q_deep if i%2 else 0.0  # only half of the injection wells recharge the lower aquifer
        model = Steady(T=T, c=c, Q=[0, Q_inj, Q3], r_out=r_out)
        r = euclidean(x, y, xgrid, ygrid)
        s += model.h(r=r.flatten()).reshape(s.shape)

    # output
    return s
```

```python
s = superposition(pwells, iwells, Q_grav, Q_pump, Q_inj, Q_deep, xgrid, ygrid)
```

Plotting the result:

```python
def contourplot(pwells, iwells, s, steps=(0.25, 0.1), lim=(0, 600), d=100, fontsize=8):

    fig, ax = plt.subplots(2, 2, figsize=(7, 7), sharex=True, sharey=True);

    ax[0][0].plot(pwells[:, 0], pwells[:, 1], 'r.', label='pumping middle layer');
    ax[0][0].plot(iwells[::2, 0], iwells[::2, 1], 'g.', label='injection middle layer');
    ax[0][0].plot(iwells[1::2, 0], iwells[1::2, 1], 'b.', label='injection middle + lower layer');
    ax[0][0].set_xlim(lim);
    ax[0][0].set_ylim(lim);
    ax[0][0].set_aspect('equal', 'box');
    ax[0][0].grid();

    for i, a in enumerate([ax[0][1], ax[1][0], ax[1][1]]):
        step = steps[0] if i == 1 else steps[1]
        cs = a.contour(xgrid, ygrid, s[i, :, :], levels=np.arange(-2, 2, step))
        a.clabel(cs, cs.levels, inline=True);
        a.set_title('layer ' + str(i+1), fontsize=fontsize);
        a.set_xlim(lim);
        a.set_ylim(lim);
        a.set_aspect('equal', 'box');
        a.grid()

    ax[0][0].set_yticks(np.arange(lim[0], lim[1] + d, d));
    ax[0][0].set_yticklabels(np.arange(lim[0], lim[1] + d, d), fontsize=fontsize);
    ax[1][0].set_yticks(np.arange(lim[0], lim[1] + d, d));
    ax[1][0].set_yticklabels(np.arange(lim[0], lim[1] + d, d), fontsize=fontsize);
    ax[1][0].set_xticks(np.arange(lim[0], lim[1] + d, d));
    ax[1][0].set_xticklabels(np.arange(lim[0], lim[1] + d, d), fontsize=fontsize);
    ax[1][1].set_xticks(np.arange(lim[0], lim[1] + d, d));
    ax[1][1].set_xticklabels(np.arange(lim[0], lim[1] + d, d), fontsize=fontsize);
    ax[0][0].set_ylabel('y (m)', fontsize=fontsize);
    ax[1][0].set_ylabel('y (m)', fontsize=fontsize);
    ax[1][0].set_xlabel('x (m)', fontsize=fontsize);
    ax[1][1].set_xlabel('x (m)', fontsize=fontsize);
    ax[0][0].legend(fontsize=fontsize);

    return ax
```

## Using TimML to simulate the drainage system

Simulating the drainage system is very straightforward using TimML.

First, we build te model by defining the aquifer system:

```python
model = timml.ModelMaq(
    z=[5, 4, 3, 2, 1, 0],  # set layer thicknesses to unity, and also include resistance layers
    kaq=T,                 # we use T as the layer thicknesses are set to unity
    c=c,                   # the resistances of the semi-pervious layers
)
```

Then, we add the pumping and injection wells using `for`-loops:

```python
# adding the pumping wells
for xy in pwells:
    x, y = tuple(xy)
    timml.Well(model, xw=x, yw=y, Qw=-Q_grav, layers=0)
    timml.Well(model, xw=x, yw=y, Qw=-Q_pump, layers=1)

# adding the injection wells
for i, xy in enumerate(iwells):
    x, y = tuple(xy)
    timml.Well(model, xw=x, yw=y, Qw=-Q_inj, layers=1)
    if i%2:
        timml.Well(model, xw=x, yw=y, Qw=-Q_deep, layers=2)
```

We also need to define a constant-head condition for point `(x0,y0)` which is at a very large distance of the wells:

```python
timml.Constant(model, xr=x0, yr=y0, layer=0);
```

All elements are added, so we are ready to solve the model:

```python
model.solve()
```

Finally, we calculate the drawdown for each node in the grid:

```python
s_tim = np.array([[model.head(x, y) for x in coord] for y in coord])  # shape is (n, n, nlay)
s_tim = np.transpose(s_tim, (2, 0, 1))  # shape is (nlay, n, n)
```

As expected, we get the same result:

```python
contourplot(pwells, iwells, s_tim);
```

## Optimizing the pumping rate using PuLP

In this notebook, we use [PuLP](https://coin-or.github.io/pulp/) to formulate and solve the linear programming problem associated with optimizing the drainage system. Although Python’s `scipy.optimize.linprog` is perfectly suitable for small, well‑structured LP problems, PuLP offers several practical advantages when building optimization models in a clear and flexible way.

PuLP allows us to express decision variables, constraints, and objective functions using a clean, symbolic, and highly readable syntax. It automatically generates the underlying constraint matrices, supports a wide range of linear solvers, and makes the model structure transparent. This makes PuLP an excellent choice when the LP model becomes more complex or when you want to maintain a clear separation between the mathematical formulation and the underlying solver mechanics.

```python
from pulp import *
```

We define the control points at which drawdown must be at least 1 m. These are the constraints.

```python
# control points
xc = [300, 300, 350, 350]
yc = [200, 250, 200, 250]
```

Since the system involves four pumping and injection rates, we must compute the head change associated with a unit discharge for each one.

First, we calculate `u_grav`, which is the head change with a unit discharge through the gravel packs:

```python
# initializing the head change vector
u_grav = np.zeros(len(xc))

# calculate drawdown due to unit discharge through gravel packs
for xy in pwells:
    x, y = tuple(xy)
    r_out = euclidean(x, y, x0, y0)  # determine the outer model boundary using (x0, y0)
    model = Steady(T=T, c=c, Q=[1, 0, 0], r_out=r_out)
    r = euclidean(x, y, xc, yc)
    u_grav += model.h(r=r)[0]  # only first layer needed
```

Second, we calculate `u_pump`, the head change caused by a unit discharge from the pumping wells:

```python
# initializing the head change vector
u_pump = np.zeros(len(xc))

# calculate drawdown due to unit discharge of pumping wells
for xy in pwells:
    x, y = tuple(xy)
    r_out = euclidean(x, y, x0, y0)  # determine the outer model boundary using (x0, y0)
    model = Steady(T=T, c=c, Q=[0, 1, 0], r_out=r_out)
    r = euclidean(x, y, xc, yc)
    u_pump += model.h(r=r)[0]  # only first layer needed
```

Third, we calculate `u_inj`, the head change caused by a unit discharge from the injection wells in the middle layer:

```python
# initializing the head change vector
u_inj = np.zeros(len(xc))

# calculate the head change due to unit recharge of injection wells (middle layer)
for i, xy in enumerate(iwells):
    x, y = tuple(xy)
    r_out = euclidean(x, y, x0, y0)
    model = Steady(T=T, c=c, Q=[0, 1, 0], r_out=r_out)
    r = euclidean(x, y, xc, yc)
    u_inj += model.h(r=r)[0]  # only first layer needed
```

Finally, we calculate `u_deep`, the head change caused by a unit discharge from the injection wells in the lower layer:

```python
# initializing the head change vector
u_deep = np.zeros(len(xc))

# calculate the head change due to unit recharge of deep injection wells (lower layer)
for i, xy in enumerate(iwells):
    if i%2: # only half of the injection wells recharge the lower aquifer
        x, y = tuple(xy)
        r_out = euclidean(x, y, x0, y0)
        model = Steady(T=T, c=c, Q=[0, 0, 1], r_out=r_out)
        r = euclidean(x, y, xc, yc)
        u_deep += model.h(r=r)[0]  # only first layer needed
```

### Scenario 1

In the first scenario, we **only optimize the pumping rate** while keeping the injection rate of the wells infiltrating the lower layer equal to -40 m³/d. Recall that the injection rate of the wells infiltrating the middle layer is determined by balancing the total injection rate with the total pumping rate.

```python
# instantiating LpProblem object (linear programming problem)
prob = LpProblem("Scenario_1", LpMinimize)  # it's a minimization problem

# defining the variables
Q_pump1 = LpVariable("Q_pump", lowBound=0)  # pumping rate middle layer (Q > 0)
Q_inj1 = LpVariable("Q_inj", upBound=0)  # injection rate middle layer (Q < 0)

# defining the objective function
prob += Q_pump1, "minimize total pumping rate"

# adding constraint Q_out == Q_in
prob += npw * (Q_grav + Q_pump1) + niw * Q_inj1 + niw / 2 * Q_deep == 0

# add constraint at the 4 control points
for i in range(len(xc)):
    # drawdown in top layer must be at least 1 m:
    prob += Q_grav * u_grav[i] + Q_pump1 * u_pump[i] + Q_inj1 * u_inj[i] + Q_deep * u_deep[i] >= 1.0

# solving the problem
print(prob.solve())
print(LpStatus[prob.status])  # checking the status of the solution
print(Q_pump1.value(), Q_inj1.value(), Q_deep)  # checking the optimized variables
```

Note that the optimal pumping rate for this scenario is more than three times higher than the initial estimate!

Checking the constraints:

```python
print(Q_grav * u_grav + Q_pump1.value() * u_pump + Q_inj1.value() * u_inj + Q_deep * u_deep)
print(npw * (Q_grav + Q_pump1.value()) + niw * Q_inj1.value() + niw / 2 * Q_deep)
```

The drawdown at the control points is at least 1 m, as required, and the sum of all pumping and injection rates is very close to zero.

Let's visualize the result:

```python
s = superposition(pwells, iwells, Q_grav, Q_pump1.value(), Q_inj1.value(), Q_deep, xgrid, ygrid)
contourplot(pwells, iwells, s, steps=(0.5, 0.2));
```

We observe that the drawdown in the top layer at the excavation site (marked by red dots) is at least 1 m.

### Scenario 2

In the second scenario, we **optimize both the pumping rate and the injection rate** of the wells infiltrating the lower layer:

```python
# instantiating LpProblem object (linear programming problem)
prob = LpProblem("Scenario_2", LpMinimize)  # it's a minimization problem

# defining the variables
Q_pump2 = LpVariable("Q_pump", lowBound=0)  # pumping rate middle layer (Q > 0)
Q_inj2 = LpVariable("Q_inj", upBound=0)  # injection rate middle layer (Q < 0)
Q_deep2 = LpVariable("Q_deep", upBound=0)  # injection rate lower layer (Q < 0)

# defining the objective function
prob += Q_pump2, "minimize total pumping rate"

# adding constraint Q_out == Q_in
prob += npw * (Q_grav + Q_pump2) + niw * Q_inj2 + niw / 2 * Q_deep2 == 0

# add constraint at the 4 control points
for i in range(len(xc)):
    # drawdown in top layer must be at least 1 m: P1*s1 + P2*s2 + P3*s3 + P4*s4 >= 1
    prob += Q_grav * u_grav[i] + Q_pump2 * u_pump[i] + Q_inj2 * u_inj[i] + Q_deep2 * u_deep[i] >= 1.0

# solving the problem
print(prob.solve())
print(LpStatus[prob.status])  # checking the status of the solution
print(Q_pump2.value(), Q_inj2.value(), Q_deep2.value())  # checking the optimized variables
```

A bit surprising here is that the linear programming solution suggests not injecting into the middle layer at all!

Let's check the constraints:

```python
print(Q_grav * u_grav + Q_pump2.value() * u_pump + Q_inj2.value() * u_inj + Q_deep2.value() * u_deep)
print(npw * (Q_grav + Q_pump2.value()) + niw * Q_inj2.value() + niw / 2 * Q_deep2.value())
```

Again, the drawdown at the control points is at least 1 m, and the sum of all pumping and injection rates is very close to zero.

Let’s generate the contour plot once more:

```python
s = superposition(pwells, iwells, Q_grav, Q_pump2.value(), Q_inj2.value(), Q_deep2.value(), xgrid, ygrid)
contourplot(pwells, iwells, s, steps=(0.25, 0.2));
```

We observe that the drawdown in the top layer at the excavation site (marked by the red dots) is at least 1 m, while the optimal pumping rate is significantly smaller than in Scenario 1!
