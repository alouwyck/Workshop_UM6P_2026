# **Part 8 - Metaheuristics for Groundwater Management**

## Introduction

In the previous notebook, we used linear programming to optimize pumping and injection rates for **aquifer reinjection**. We applied this approach to two illustrative cases:

1. a simple system consisting of one pumping well and one injection well, and
2. a more advanced construction dewatering scenario in which an excavation pit is surrounded by multiple pumping and injection wells that reinject the extracted groundwater into the aquifer.

In this notebook, we revisit these same examples, but we approach the optimization task from a different angle. Instead of relying on linear superposition and linear constraints, we explore **metaheuristic optimization methods**—specifically, **Simulated Annealing** (SA) and a **Genetic Algorithm** (GA), both available in `scipy.optimize`. Metaheuristics are global optimization techniques that do not require linearity or convexity, making them highly flexible and capable of handling more complex or nonlinear simulation models.

Our goal is to demonstrate how these methods can be applied to **optimize pumping rates** in aquifer reinjection systems, how they compare to linear programming, and in what situations they may offer clear advantages. During the second day of the workshop, we will explore in detail how these metaheuristic methods work.

## Mathematical Formulation

Recall that the goal is to **minimize the total pumping rate**:

> $\min \sum\limits_i c_i Q_i$

with $Q_i$ the pumping rate [L³/T] of well $i$ and $c_i$ the coefficients that weight the contribution of each well $i$. We define these coefficients as:

> $
c_i =
\begin{cases}
1 & \text{if well } i \text{ is a pumping well} \\
0 & \text{if well } i \text{ is a reinjection well}
\end{cases}
$

This minimization problem is subject to the following **constraints** at several control points $j$:

> $s_j^{\text{min}} \leq s_j \leq s_j^{\text{max}}$

with $s$ the head change [L].


For **metaheuristic algorithms** like Genetic Algorithms (GA) or Simulated Annealing (SA), we typically formulate the problem as a single objective function to minimize. This function $f$, often called a **cost function** or **fitness function**, combines the primary **objective** (i.e., minimize total pumping) with **penalties** for violating the constraints:

> $f = \sum\limits_i c_iQ_i + \lambda \sum\limits_j \max\big(0, a(s_j - s_j^{\text{lim}}) \big)$
>
where
- $Q_i$ is the pumping rate [L³/T] of well $i$
- $c_i$ is the coefficient that weights the contribution of well $i$
- $\lambda$ is the penalty weight, chosen sufficiently large to strongly enforce the constraints
- $s_j$ is the head change [L] at control point $j$
- $s_j^{\text{lim}}$ is the prescribed head change limit ($s_j^{\text{min}}$ or $s_j^{\text{max}}$) at control point $j$
- $a$ is the inequality indicator, defined as

$$
a =
\begin{cases}
  1, & \text{if } s_j \le s_j^{\text{lim}}, \\
 -1, & \text{if } s_j \ge s_j^{\text{lim}}.
\end{cases}
$$

Once the cost or fitness function is defined, metaheuristic algorithms can explore the **solution space** to identify the parameter values that yield the lowest cost or highest fitness. They typically do not guarantee the exact optimum, but they are very effective at finding high‑quality **approximations** within a reasonable amount of time.

## OO Python

In this section, we first prepare the code that we'll need. We import the `numpy` library and the `pyplot` module, and define the `thiem` function as well as the `Well` and `Model` classes that we implemented in the previous notebook.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def thiem(Q, T, R, r):
    '''
    Simulate drawdown using the Thiem equation
    Q: pumping rate [L³/T] - float
    T: aquifer transmissivity [L²/T] - float
    R: radius of influence [L] - float
    r: radial distance [L] - array-like
    returns drawdown s [L] for each distance r
    '''
    return Q/2/np.pi/T * np.log(np.array(r)/R)
```

```python
class Well:
    """
    A pumping or injection well represented by the Thiem equation.

    Parameters
    ----------
    model : Model
        The Model object to which the well belongs.
    Q : float
        Pumping rate [L³/T]. Positive for pumping, negative for injection.
    xw, yw : float
        Coordinates [L] of the well.
    rw : float, optional
        Well radius [L]. Distances smaller than this are set to rw
        to avoid singularities. Default is 0.1 m.
    """

    def __init__(self, model, Q, xw, yw, rw=0.1):
        self.model = model
        self.Q = Q
        self.xw, self.yw = xw, yw
        self.rw = rw

    def s(self, x, y):
        """
        Compute drawdown at coordinates (x, y) due to this well
        using the Thiem equation.

        Parameters
        ----------
        x, y : float or array-like
            Coordinates [L] at which the drawdown is evaluated.

        Returns
        -------
        ndarray or float
            Drawdown value(s) [L] at the given coordinate(s).
        """
        x, y = np.array(x), np.array(y)
        r = np.sqrt((x - self.xw)**2 + (y - self.yw)**2)
        if x.ndim: r[r < self.rw] = self.rw  # prevent singularity
        elif r < self.rw: r = self.rw

        return thiem(self.Q, self.model.T, self.model.R, r)
```

```python
class Model:
    """
    Model for single homogeneous aquifer represented by transmissivity, initial head,
    and a list of wells.

    Parameters
    ----------
    T : float
        Aquifer transmissivity [L²/T].
    h0 : float, optional
        Initial hydraulic head [L] in the aquifer. Default is 0.
    R : float, optional
        Approximate distance [L] to constant-head point for wells in the aquifer.
        Default is 1e6.
    """

    def __init__(self, T, h0=0, R=1e6):
        self.T = T
        self.h0 = h0
        self.R = R
        self.wells = []

    def add_well(self, Q, xw, yw, rw=0.1):
        """
        Add a pumping or injection well to the model.

        Parameters
        ----------
        Q : float
            Pumping rate [L³/T].
        xw, yw : float
            Well coordinates [L].
        rw : float, optional, default is 0.1
            Well radius [L].

        Returns
        -------
        Well
            The newly created Well object.
        """
        well = Well(self, Q, xw, yw, rw)
        self.wells.append(well)
        return well

    def h(self, x, y):
        """
        Compute the hydraulic head at coordinates (x, y),
        obtained by superposition of all wells.

        Parameters
        ----------
        x, y : float or array-like
            Coordinates [L] at which the head is evaluated.

        Returns
        -------
        ndarray or float
            Hydraulic head [L] value(s) at the given coordinate(s).
        """
        return self.h0 + np.sum([well.s(x, y) for well in self.wells], axis=0)
```

## Pumping and Reinjection Well System

Let's start again with the simple two-well system:
- the distance between pumping and injection well is 50 m;
- the pumping rate is 3000 m³/d;
- the aquifer has a transmissivity of 3000 m²/d.

```python
T = 100   # aquifer transmissivity (m²/d)
Q = 3000  # pumping rate (m³/d)

# well coordatines
x_pump, y_pump = 25, 0  # pumping well
x_inj, y_inj = -25, 0   # injection well

# grid
x, y = np.meshgrid(np.linspace(-100, 100, 100),  # x-axis: 100 point between -100 and 100
                   np.linspace(-50, 50, 100))    # y-axis: 100 point between -50 and 50
```

Building the model and simulate the drawdowns:

```python
model = Model(T)                          # model
pump = model.add_well(Q, x_pump, y_pump)  # pumping well
inj = model.add_well(-Q, x_inj, y_inj)    # injection well
s = model.h(x, y)                         # simulate drawdowns
```

Visualizing the result using a contour plot:

```python
plt.figure(figsize=(10, 5))
plt.contourf(x, y, s, levels=np.arange(-25, 26, 1))
plt.colorbar()
ax = plt.gca() # Get the current axes
ax.set_aspect('equal') # Set aspect ratio to equal
```

Computing the drawdown at point (10, 0):

```python
x_target, y_target = 10, 0
print(model.h(x_target, y_target))
```

Suppose we now want the drawdown at point (10, 0) to be equal to -2 m. The **analytical solution** is calculated as follows:

```python
s_target = -2

r_pump = np.sqrt((x_target - x_pump)**2 + (y_target - y_pump)**2)
r_inj = np.sqrt((x_target - x_inj)**2 + (y_target - y_inj)**2)

Q_opt = 2*np.pi*T*s_target / np.log(r_pump/r_inj)
print('optimal pumping rate:', Q_opt, 'm³/d')
```

Rebuilding the model using the derived optimal pumping rate and verifying the constraint:

```python
model_opt = Model(T)                              # model
pump = model_opt.add_well(Q_opt, x_pump, y_pump)  # pumping well
inj = model_opt.add_well(-Q_opt, x_inj, y_inj)    # injection well
print(model_opt.h(x_target, y_target).item())     # drawdown at target point
```

Now we show how to solve this optimization problem using a simple **genetic algorithm** (GA). This is a **metaheuristic** optimization method inspired by natural evolution.

It works by generating a *population* of candidate solutions and improving them over successive *generations*. At each iteration, the best-performing solutions are *selected*, combined (*crossover*), and randomly modified (*mutation*) to explore the *search space*. Over time, this evolutionary process drives the population toward increasingly better solutions.

The objective is still to minimize the pumping rate:

> $\min Q$

The hard constraint at point (10, 0)

> $s_{\text{target}} = -2$

needs to be formulated as a **soft constraint** or penalty:

> $\min |s - s_{\text{target}}|$

Combining these two gives the **fitness function** to minimimze:

> $\min f$

with:

> $f = Q + \lambda |s - s_{\text{target}}|$

where $\lambda$ is the penalty weight.

Let's define a `fitness` function that implements this:

```python
# fitness function
def fitness(Q):
    model = build_model(Q)
    penalty = abs(model.h(x_target, y_target) - s_target)
    lambda_ = 1e6
    return Q + lambda_ * penalty  # large penalty weight

# function to build model
def build_model(Q):
    model = Model(T)
    model.add_well(Q, x_pump, y_pump)
    model.add_well(-Q, x_inj, y_inj)
    return model
```

Function `genetic_algorithm` implements the simple GA:

```python
# GA function
def genetic_algorithm(fitness, pop_size, num_of_gen, mut_std, seed=None):
    """
    fitness: fitness function (lambda Q: f(Q))
    pop_size: population size (int)
    num_of_gen: number of generations (int)
    mut_std: mutation standard deviation (float)
    seed: seed for numpy random generator (int)
    """
    # set seed
    if seed: np.random.seed(seed)

    # initialize population randomly
    population = np.random.uniform(0, 10_000, pop_size)
    best_scores = []

    for generation in range(num_of_gen):
        # evaluate fitness
        scores = np.array([fitness(Q) for Q in population])
        best_scores.append(np.min(scores))

        # select best half (= survival of the fittest)
        idx = np.argsort(scores)
        population = population[idx[:pop_size // 2]]

        # crossover + mutation
        children = []
        for _ in range(pop_size // 2):
            # crossover: select two random parents and average to get child
            parents = np.random.choice(population, 2)
            child = np.mean(parents)
            # mutation: add random gaussian noise
            child += np.random.normal(0, mut_std)
            child = max(0, child)  # enforce Q >= 0
            # new generation
            children.append(child)

        # population = survivors + new children
        population = np.concatenate([population, children])

    # pick best solution from final population
    Q_opt = population[np.argmin([fitness(Q) for Q in population])]
    return Q_opt, best_scores
```

Let's apply functions `fitness` and `genetic_algorithm` to find the optimal pumping rate `Q_opt`:

```python
Q_opt, best_scores = genetic_algorithm(fitness, pop_size=20, num_of_gen=25, mut_std=10, seed=42)
print('optimal pumping rate:', Q_opt, 'm³/d')
model_opt = build_model(Q_opt)
print('drawdown at point (10, 0):', model_opt.h(x_target, y_target), 'm')
```

We obtain a very accurate **approximation** of the optimal pumping rate.

Plotting the best score (i.e., the optimal pumping rate) across generations clearly shows that the GA converges:

```python
plt.semilogy(best_scores)
plt.xlabel("population")
plt.ylabel("best score")
plt.grid();
```

 Whether the algorithm converges, and how fast, depends on the population size `pop_size`, the number of generations `num_of_gen`, and the mutation standard deviation `mut_std`.

Here, we implemented a very simple version of a GA because the optimization problem we are solving is also quite straightforward. It also serves as a nice example to highlight the similarities and differences with linear programming. Metaheuristics become useful particularly for nonlinear problems or search spaces that are too large or complex for classical optimization methods.

## A More Challenging Example

The example consists of a rectangular excavation of 50 m × 30 m, centered in the model domain. To enable dry working conditions, groundwater levels inside the excavation are lowered using a series of pumping wells placed around the site.

The pumping wells are arranged in a rectangular pattern at a distance of 10 m from the excavation boundary, whereas the distance between two adjacent wells is also 10 m. This creates a controlled drawdown that keeps the groundwater level below the excavation base.

To reduce the impact of dewatering on the surrounding area, the extracted groundwater is re-injected into the aquifer using a second ring of wells located further away. These injection wells are positioned at a distance of 25 m from the excavation boundary, forming a wider perimeter around the site.

The purpose of the reinjection is to compensate for the extracted water, thereby limiting drawdown outside the construction area and maintaining the regional groundwater balance.

We consider a homogeneous aquifer with constant thickness equal to 20 m and horizontal conductivity equal to 10 m/d:

```python
T = 20 * 10  # transmissivity T = KD
```

We had to decouple the NS‑ and EW‑oriented pumping and injection wells to minimize the total pumping rate and to satisfy the following constraints:

- a minimum drawdown of -2 m at the center of the excavation;
- no negative drawdown anywhere at 100 m;
- no head rise greater than 2 m at the injection wells.

In this new configuration, the NS‑oriented pumping wells all operate at the same pumping rate, and the water they extract is reinjected through the NS‑oriented injection wells. The same principle applies to the EW‑oriented wells.

```python
# Excavation boundary (brown)
excavation = {"xmin": -25, "xmax": 25, "ymin": -15, "ymax": 15}
```

```python
# Pumping wells (blue)

# Top and bottom rows (EW-oriented wells)
pumping_wells_EW = []
for x in np.arange(-25, 30, 10):
    pumping_wells_EW.append((x, 25))
    pumping_wells_EW.append((x, -25))

# Left and right columns (NS-oriented wells)
pumping_wells_NS = []
for y in np.arange(-15, 20, 10):
    pumping_wells_NS.append((-35, y))
    pumping_wells_NS.append((35, y))

# All pumping wells
pumping_wells = pumping_wells_EW + pumping_wells_NS
```

```python
# Injection wells (green)

# Top and bottom rows (EW-oriented wells)
injection_wells_EW = []
for x in [-25, 0, 25]:
    injection_wells_EW.append((x, 40))
    injection_wells_EW.append((x, -40))

# Left and right columns (NS-oriented wells)
injection_wells_NS = []
for y in [-15, 0, 15]:
    injection_wells_NS.append((-50, y))
    injection_wells_NS.append((50, y))

# All injection wells
injection_wells = injection_wells_EW + injection_wells_NS
```

```python
# Function to plot the excavation and the wells
def plot_wells():

    fig, ax = plt.subplots(figsize=(6,6))

    # Excavation
    rect = plt.Rectangle(
        (excavation["xmin"], excavation["ymin"]),
        excavation["xmax"] - excavation["xmin"],
        excavation["ymax"] - excavation["ymin"],
        color="brown"
    )
    ax.add_patch(rect)

    # Wells
    px, py = zip(*pumping_wells)
    ix, iy = zip(*injection_wells)

    ax.scatter(px, py, color="blue")   # pumping
    ax.scatter(ix, iy, color="green")   # injection

    # Limits
    xmin, xmax = -60, 60
    ymin, ymax = -50, 50
    ax.set_xlim(xmin, xmax)
    ax.set_ylim(ymin, ymax)

    # ---- TICKS ----
    # Major ticks every 10 m (with labels)
    ax.set_xticks(np.arange(xmin, xmax + 10, 10))
    ax.set_yticks(np.arange(ymin, ymax + 10, 10))

    # Minor ticks every 5 m (no labels)
    ax.set_xticks(np.arange(xmin, xmax + 5, 5), minor=True)
    ax.set_yticks(np.arange(ymin, ymax + 5, 5), minor=True)

    # ---- GRID ----
    ax.grid(which='minor', linewidth=0.5)
    ax.grid(which='major', linewidth=1.0)

    # Keep square cells
    ax.set_aspect('equal')

    # Return axes
    return ax

# Plot
plot_wells();
```

First, we state the **optimization problem** mathematically.

The objective is minimize the total pumping rate:
> $$\min \big( Q_{\text{EW}} + Q_{\text{NS}} \big)$$
>
The decision variables are the pumping rates of the EW- and NS-oriented pumping wells:
> $$Q_{\text{EW}} \ge 0$$
> $$Q_{\text{NS}} \ge 0$$
>
There is a constraint at the center of the excavation:
> $$s_{\text{center}} \le -2$$
>
There is also a constraint at 100 m:
> $$s_{\text{100m}} \ge 0$$
>
Finally, there is a constraint at the injections wells:
> $$s_{\text{inj}} \le 2$$

This problem is translated to the following **fitness function**:

> $f = Q_{\text{EW}} + Q_{\text{NS}} + \lambda \bigg[ \max(0, s_{\text{center}}+2) + \max(0, -s_{\text{100m}}) + \max(0, s_{\text{inj}}-2) \bigg]$

The metaheuristic algorithm aims to minimize the fitness function $f$. Note that the constraints are incorporated as penalty terms that are added to the objective function. A large penalty weight $\lambda$ ensures that constraint violations are heavily penalized, guiding the optimizer towards feasible solutions.



Let's implement these constraints:

```python
# constraint at center of excavation
xy_center = (0, 0)
s_center = -2

# constraint at 100 m
xy1_100 = (100, 0)
xy2_100 = (0, 100)
s_100 = 0

# constraint at injection wells:
xy1_inj = (0, 40)
xy2_inj = (50, 0)
s_inj = 2

# dictionary
CONSTRAINTS = dict(
    xy=np.array(        [xy_center, xy1_100, xy2_100, xy1_inj, xy2_inj]),
    s_lim=np.array(     [s_center,  s_100,   s_100,   s_inj,   s_inj  ]),
    inequality=np.array([1,         -1,      -1,      1,       1      ])  # 1 is <= and -1 is >=
)
```

For convenience, we have grouped all constraints in a single Python dictionary. The field `inequality` is the inequality indicator $a$, which equals 1 for a ‘less‑than‑or‑equal‑to’ constraint and −1 for a ‘greater‑than‑or‑equal‑to’ constraint.

For convenience, we also define a function `build_model` that builds the model simulating the dewatering system, using the pumping rates $Q_{\text{EW}}$ and $Q_{\text{NS}}$, which are passed to the function as a vector `Q_vector`:


```python
def build_model(Q_vector):
    # Q_vector is [Q_EW, Q_NS]
    Q_pump_EW, Q_pump_NS = Q_vector
    Q_inj_EW = -len(pumping_wells_EW) * Q_pump_EW / len(injection_wells_EW)
    Q_inj_NS = -len(pumping_wells_NS) * Q_pump_NS / len(injection_wells_NS)
    model = Model(T)
    for pumping_well in pumping_wells_EW:
        model.add_well(Q_pump_EW, *pumping_well)
    for pumping_well in pumping_wells_NS:
        model.add_well(Q_pump_NS, *pumping_well)
    for injection_well in injection_wells_EW:
        model.add_well(Q_inj_EW, *injection_well)
    for injection_well in injection_wells_NS:
        model.add_well(Q_inj_NS, *injection_well)
    return model
```

Finally, we define the `cost_function`, which also accepts `Q_vector` as input:

```python
LAMBDA = 1e6  # a large number for the penalty weight to ensure the constraints are strongly enforced

def cost_function(Q_vector):
    # Q_vector is [Q_EW, Q_NS]

    # Ensure pumping rates are non-negative (can also be handled by bounds in optimizers)
    if np.any(Q_vector) < 0:
        return float('inf')

    # Build model using input vector Q
    model = build_model(Q_vector)

    # Compute current head change at constraints
    s = np.array([model.h(*xy) for xy in CONSTRAINTS['xy']])

    # Calculate total penalty for violating constraints
    penalty = np.sum(np.maximum(0, (s - CONSTRAINTS['s_lim']) * CONSTRAINTS['inequality']))

    # Objective: Minimize total pumping
    Q_total = np.sum(Q_vector)

    # Return total cost
    return Q_total + LAMBDA * penalty
```

In the previous notebook, we determined the optimal pumping rates by applying **linear programming**:

```python
Q_EW = 338.70339754165803 # m³/d
Q_NS = 318.09403315932605 # m³/d
```

Let’s verify the cost associated with these optimal rates:

```python
print("Linear programming cost:", cost_function(np.array([Q_EW, Q_NS])))
```

First, we apply a **Genetic Algorithm** to minimize this `cost_function`. Actually, we use `scipy.optimize.differential_evolution`, a global optimization algorithm inspired by genetic algorithms. It's suitable for minimizing functions over a multi-dimensional parameter space, especially when the function is non-convex or has many local minima.

```python
from scipy.optimize import differential_evolution

# Define bounds for Q_EW and Q_NS (e.g., 0 to 1000 m³/d)
bounds = [(0, 1000), (0, 1000)]

result_ga = differential_evolution(cost_function, bounds, maxiter=1000, popsize=15, seed=42)

if result_ga.success:
    Q_EW_ga, Q_NS_ga = result_ga.x
    print("Genetic Algorithm (Differential Evolution) Results:")
    print(f"Optimal Q EW: {Q_EW_ga:.3f} m³/d")
    print(f"Optimal Q NS: {Q_NS_ga:.3f} m³/d")
    print(f"Total Pumping (Objective): {Q_EW_ga + Q_NS_ga:.3f} m³/d")
    print(f"Final Cost: {result_ga.fun:.3f}")
else:
    print("Genetic Algorithm (Differential Evolution) failed to converge:")
    print(result_ga.message)
```

For **Simulated Annealing**, we use `scipy.optimize.dual_annealing`, another global optimization algorithm that performs well on difficult, high-dimensional global optimization problems. It's a robust method for exploring complex search spaces.

```python
from scipy.optimize import dual_annealing

# Define bounds for Q_EW and Q_NS (e.g., 0 to 1000 m³/d)
bounds = [(0, 1000), (0, 1000)]

# Tune parameters: increased maxiter
result_sa = dual_annealing(cost_function, bounds, maxiter=2000, seed=42)

if result_sa.success:
    Q_EW_sa, Q_NS_sa = result_sa.x
    print("Simulated Annealing Results:")
    print(f"Optimal Q EW: {Q_EW_sa:.3f} m³/d")
    print(f"Optimal Q NS: {Q_NS_sa:.3f} m³/d")
    print(f"Total Pumping (Objective): {Q_EW_sa + Q_NS_sa:.3f} m³/d")
    print(f"Final Cost: {result_sa.fun:.3f}")
else:
    print("Simulated Annealing failed to converge:")
    print(result_sa.message)
```

Simulating the head change at the control points inside the cost function can be done more efficiently by using the head change per unit discharge, just as we did in the linear‑programming approach. The difference is that this form of superposition is required in linear programming, whereas metaheuristic methods do not rely on it. This makes them more flexible, as they can also handle non‑linear simulation models.

Here, the most important **conclusion** is that both metaheuristic approaches find solutions very close to the optimal values obtained by linear programming, indicating their effectiveness even for problems where analytical or linear programming solutions exist. This demonstrates their utility for more complex, potentially non-linear, scenarios where traditional methods might struggle. On day 2 of this workshop, we will take a closer look at metaheuristics.
