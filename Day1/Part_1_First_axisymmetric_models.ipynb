## 1. The Thiem-Dupuit equations

The Thiem equation is used to calculate steady-state *confined* one-dimensional axisymmetric flow to a pumping well, while the Dupuit equation is used to calculate steady-state *unconfined* one-dimensional axisymmetric flow to a pumping well. Let's define a function for each equation:

```python
def thiem(r, T, Q, R):
    """
    Calculate drawdown at given distances r according to the Thiem formula for steady confined flow.

    Parameters
    ----------
    r : array_like
      Radial distances [L], which should be smaller than `R`.
    T : float
      Aquifer transmissivity [L²/T].
    Q : float
      Pumping rate [L³/T] of the well.
    R : float
      Radial distance [L] of the outer aquifer boundary.

    Returns
    -------
    s : ndarray
      Drawdown [L] at given distances `r`. The shape of `s` is the same as the shape of `r`.
    """
    r = np.array(r)
    return Q / 2 / np.pi / T * np.log(R / r)
```

```python
def dupuit(r, K, h0, Q, R):
    """
    Calculate drawdown at given distances r according to the Dupuit formula for steady unconfined flow.

    Parameters
    ----------
    r : array_like
      Radial distances [L], which should be smaller than `R`.
    K : float
      Aquifer conductivity [L/T].
    h0 : float
       Initial hydraulic head [L] which is the initial aquifer thickness before pumping.
       `h0` is also the constant head at the outer aquifer boundary at distance `R`.
    Q : float
      Pumping rate [L³/T] of the well (which is positive in case of extraction).
    R : float
      Radial distance [L] of the outer aquifer boundary.

    Returns
    -------
    s : ndarray
      Drawdown [L] at given distances `r`. The shape of `s` is the same as the shape of `r`.
    """
    return h0 - np.sqrt(h0**2 - Q / np.pi / K * np.log(R / r))
```

An example to illustrate the use of these functions:

```python
Q  = 1000    # pumping rate (m³/d): Q > 0 -> extraction
R  = 500     # radius of influence (m)
K  = 10      # conductivity (m/d)
h0 = 20      # initial head = initial aquifer thickness (m)
T  = K * h0  # aquifer transmissivity (m²/d)

r = np.logspace(-2, np.log10(R), 100)   # radial distances (m)
s1 = dupuit(r=r, Q=Q, K=K, R=R, h0=h0)  # drawdown (m) according to the Dupuit formula
s2 = thiem(r=r, Q=Q, T=T, R=R)          # drawdown (m) according to the Thiem formula
```

The corresponding plot:

```python
plt.semilogx(r, s1, 'b-', label='unconfined (Dupuit)');  # drawdown s (m) according to the Dupuit formula
plt.semilogx(r, s2, 'b:', label='confined (Thiem)');     # drawdown s (m) according to the Thiem formula
plt.xlabel('r (m)');
plt.ylabel('s (m)');
plt.legend();
plt.grid();
```

Although previous semi-logarithmic graph is more practical, it's also possible to visualize the actual cone of depression:

```python
plt.figure(figsize=(10, 2));
plt.plot(-r, -s1, 'b-');
plt.plot(r, -s1, 'b-');
plt.plot(-r, -s2, 'b:');
plt.plot(r, -s2, 'b:');
plt.grid();
```

## 2. TimML

It is also possible to use TimML to simulate steady flow to a pumping well. [TimML](https://timml.readthedocs.io/en/latest/) is a Python package for the modeling of steady-state multi-layer groundwater flow with analytic elements developed by [Mark Bakker](https://www.tudelft.nl/citg/over-faculteit/afdelingen/watermanagement/medewerker/staff-hydrology/academic-staff/profdrir-mark-bakker), Professor in Groundwater Engineering at the Technical University of Delft in the Netherlands.

First, we need to install TimML:

```python
!pip install timml -q
```

After installing the TimML package, we import it:

```python
import timml
```

The first step in a TimML simulation is to **create a model** and define the **aquifer properties**. Here we use the `ModelMaq` class and specify the hydraulic conductivity `kaq`, and the top and bottom of the aquifer, which are assigned to `z`. The `ModelMaq` class returns a `model` object:

```python
model = timml.ModelMaq(
    kaq=K,      # hydraulic conductivity (m/d)
    z=[h0, 0],  # top and bottom of aquifer (m)
)
```

Once the model is created, we can **add analytic elements** to the model. Here, we need a `Well` with pumping rate equal to `Q`. To keep it simple, we define the well at (0,0), and set the well radius `rw` to the first radial distance in array `r`:

```python
well = timml.Well(
    model=model,  # model to which element is added
    xw=0,         # x-location of well (m)
    yw=0,         # y-location of well (m)
    Qw=Q,         # discharge of well (m³/d), positive for extraction
    rw=r[0],      # well radius (m), we take the first element of r
)
```

 Recall that we also have to define a `Constant` head at distance `R`:

```python
constant = timml.Constant(
    model=model,  # model to which element is added
    xr=R,         # x-location of fixed head (m)
    yr=0,         # y-location of fixed head (m)
    hr=h0,        # fixed head (m)
)
```

Now we're ready to **solve the model**, which means that any unknown parameters in the model are computed.

```python
model.solve()
```

Calling method `model.head()` gives the head at a specified point (x,y). Drawdown at distances `r` is calculated as the difference between the initial head `h0` and the head during pumping:

```python
s3 = h0 - np.array([model.head(x=r, y=0) for r in r])
```

This gives virtually the same result as the Thiem solution:

```python
plt.semilogx(r, s2, 'k-', label='Thiem');
plt.semilogx(r, s3, 'r:', label='TimML');
plt.legend();
plt.xlabel('r (m)');
plt.ylabel('s (m)');
plt.grid();
```

Another nice feature of TimML is method `contour`:

```python
model.plots.contour(
    win=[-R/2, R/2, -R/2, R/2],  # window to contour [xmin, xmax, ymin, ymax]
    ngr=100,                     # number of points where to compute the head
    levels=np.arange(20),        # levels to be contoured
    labels=True,                 # add labels along the contours
    decimals=1,                  # print labels with 1 decimal place
    legend=True,                 # add a legend
    figsize=(5, 5),              # specify a figure size of 5 by 5 inches
);
```

Check the TimML documentation to see how you can plot pathlines and a capture zone.

Is it possible to simulate a Dupuit model using TimML? The answer is no, because TimML only supports confined layers, or more accurately, horizontal layers with constant thickness and constant hydraulic properties.

The various steps you need to take with TimML may seem somewhat cumbersome at first glance, but this is because TimML is a **generic model** for simulating groundwater flow in groundwater systems consisting of **multiple aquifers and aquitards**. In addition to wells and constant heads, **more complex elements** can also be added to a model, such as line sinks and areal sinks, which can be used to simulate interactions with rivers and infiltration areas.

For instance, we assumed there is no groundwater flow in the aquifer before the pumping starts, which is very unlikely. TimML provides a `UFlow` element to add uniform flow to the aquifer:  

```python
flow = timml.Uflow(
    model=model,  # model to which element is added
    slope=0.01,   # hydraulic gradient = head difference / distance (in direction of flow)
    angle=0,      # direction of uniform flow, straight East is zero degrees
)
```

**Uniform flow** refers to a situation where groundwater moves through the aquifer at a constant velocity and in a single direction, typically caused by a regional hydraulic gradient. In other words, the head decreases linearly across the aquifer, and flow lines are parallel.

We can calculate the flow rate $q$ of this uniform flow using Darcy's law. The `slope` input parameter of the `Uflow` element actually defines the **hydraulic gradient** $\Delta h / \Delta x$, which is set to `0.01` in this case.

Let's solve the model again so it accounts for the added uniform flow:

```python
model.solve()
```

The contour plot looks different now:

```python
model.plots.contour(
    win=[-R, R, -R, R],        # window to contour [xmin, xmax, ymin, ymax]
    ngr=100,                   # number of points where to compute the head
    levels=np.arange(20, 31),  # levels to be contoured
    labels=True,               # add labels along the contours
    decimals=1,                # print labels with 1 decimal place
    legend=False,              # don't add a legend
    figsize=(5, 5),            # specify a figure size of 5 by 5 inches
);
```

Remember that we defined a constant head of 20m at a point 500m east of the pumping well along the x-axis. As expected, the hydraulic head there is indeed 20m. There is a uniform flow increasing from east to west, so at 500m west of the well the head is 30m. This is easy to explain because the hydraulic gradient is 0.01, meaning the head rises by 1cm per meter, or 10m per kilometer. Uniform flow normally produces parallel streamlines — in this case running north to south — but the pumping well disturbs this pattern. At the well location (0,0), we clearly see a local cone of depression.

As already mentioned, it is also possible to include the effect of a river or a canal in the model by using, for example, the `River` element. Here we add an **east-west oriented canal** located 100m north of the pumping well. The `River` element represents the canal as a straight line between two points, `(x1, y1)` and `(x2, y2)`. The canal has a fixed water level `hls`, which we will set to 20m. We can also specify a width `wh`; here we take 10m. Finally, we define a hydraulic resistance `res`. Simply put, this resistance controls the exchange of water between the canal and the aquifer, and the higher the resistance, the more difficult it is for water to move between them.

```python
canal = timml.linesink.River(
    model,            # model to which element is added
    x1=-500, y1=100,  # coordinates of the first point of the canal
    x2=500, y2=100,   # coordinates of the second point of the canal
    hls=20.0,         # constant water level (m) in the canal
    wh=10,            # width (m) of the canal
    res=10,           # hydraulic resistance (d) between canal and aquifer
)
```

Solving the model to take into account the newly added element:

```python
model.solve()
```

The contour plot clearly shows the draining effect of the canal:

```python
model.plots.contour(
    win=[-R, R, -R, R],        # window to contour [xmin, xmax, ymin, ymax]
    ngr=100,                   # number of points where to compute the head
    levels=np.arange(20, 31),  # levels to be contoured
    labels=True,               # add labels along the contours
    decimals=1,                # print labels with 1 decimal place
    legend=False,              # don't add a legend
    figsize=(5, 5),            # specify a figure size of 5 by 5 inches
);
```

Recall that the contour plot shows lines of equal hydraulic head. To visualize groundwater movement within the aquifer, we add so-called **pathlines** to the plot. These are shown in orange and are perpendicular to the contour lines, indicating the direction of groundwater flow.

```python
# contour plot
model.plots.contour(
    win=[-R, R, -R, R],        # window to contour [xmin, xmax, ymin, ymax]
    ngr=100,                   # number of points where to compute the head
    levels=np.arange(20, 31),  # levels to be contoured
    labels=True,               # add labels along the contours
    decimals=1,                # print labels with 1 decimal place
    legend=False,              # don't add a legend
    figsize=(5, 5),            # specify a figure size of 5 by 5 inches
)

# pathlines
xstart = np.hstack((
    np.arange(-400, 501, 100),
    -500 * np.ones(21),
    np.arange(-400, 501, 100),
))
ystart = np.hstack((
    500 * np.ones(10),
    np.arange(-500, 501, 50),
    -500 * np.ones(10),
))
model.plots.tracelines(
    xstart=xstart,                 # x-locations of starting points
    ystart=ystart,                 # y-locations of starting points
    zstart=np.zeros(len(xstart)),  # z-locations of starting points
    hstepmax=20,                   # maximum horizontal step size
    color="C1",                    # color of lines
);
```

Remember that groundwater flows from areas of high hydraulic head to areas of low head. Here, this is clearly from west to east because of the uniform flow element we've added to the model. Moreover, the pathlines clearly reveal that groundwater also flows toward the canal, meaning the canal drains the aquifer. And of course, water flows toward the pumping well as well.
